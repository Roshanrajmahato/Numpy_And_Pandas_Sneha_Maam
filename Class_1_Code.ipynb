## Row Selection And column Selection
### .loc[] Attribute is used to perform that
### .loc[row_based , col_based] 
### .loc[Row Selection , column Selection]
- Row Selection or row_based 
     - index
     - list_of_index
     - slicing
     - condition

**These Things are applied in First Argument of .loc[] Attribute**

- column Selection or col_based
  - col_name
  - list_of_cols
  
**These Things are applied in Second Argument of .loc[] Attribute**

```python
# .loc[7] To get the record of 7th row (Row Selection)
df.loc[7]
```

```python
# .loc[[5,6]] To get the record of 5th or 6th row (Row Selection)
df.loc[[5,6]]
```

```python
# df.loc[[5,6],'Food'] To get the record of 5th or 6th row (Row Selection) and only one 'Food' column (Column Selection)
df.loc[[5,6],'Food']
```

```python
# .loc[[5,6],['Food','Rating']] To get the record of 5th or 6th row (Row Selection) From 'Food' 'Rating' Column  (Column Selection)
df.loc[[5,6],['Food','Rating']]
```

```python
df.loc[7,'Food']
```

```python
df.loc[(df['Price']>200)]
```

```python
df.loc[(df['Price']>200) & (df['Rating']==9)]
```

```python
df.loc[(df['Price']>200) & (df['Rating']==9) ,'Food']

```

```python
df.loc[(df['Price']>200) & (df['Rating']==9) ,['Food','Rating']]
```
