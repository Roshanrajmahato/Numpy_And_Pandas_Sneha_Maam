# Concatenation
.concat([DataFrame1,DataFrame2,DataFrame3,DataFrame4,......],axis=1)
axis=0 is for joining rows data or record or value  \n
axis=1 is for joining column data or record or value

```python
new_df=pd.concat([df,df1],axis=1)
```

```python
new_df
```

```python
df=pd.concat([new_df,df2],axis=0)
```
