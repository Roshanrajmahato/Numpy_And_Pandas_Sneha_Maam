```python
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split,GridSearchCV
# from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score # Evalutions For Regrassion
```

```python
df=load_diabetes(as_frame=True)
X=df.data
y=df.target
```

```python
X
```

```python
y
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
model=RandomForestRegressor(n_estimators=100)
```

```python
grid_search=GridSearchCV(
    estimator=RandomForestRegressor(random_state=42),
    param_grid={
        'criterion':['squared_error','absolute_error'],
        'n_estimators':[50,100,200],
        'max_depth':[None,5,10,50],
        'min_samples_split':[2,5,7,10],
        'min_samples_leaf':[1,3,5,10],
        'max_features':['sqrt',1.0] 
    },cv=5,verbose=5
)
grid_search.fit(Xtrain,ytrain)
```

```python
grid_search.score(Xtrain,ytrain)
```

```python
grid_search.score(Xtest,ytest)
```
