```python
df.columns
```

```python
df['total score']=(df['math score']+df['reading score']+df['writing score'])/3
```

```python
df=df.drop(['math score','reading score','writing score'],axis=1)
```

```python
df
```

```python

X=df.iloc[:,:-1]
y=df['total score']
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
# OneHotEncoder()
encoder=OneHotEncoder(sparse_output=False,handle_unknown='ignore') 
# sparse_output=False :- Directly convert generative object to array
# handle_unknown='ignore' :- To handle unkown value
```

```python
train_encoder_values=encoder.fit_transform(Xtrain)
```

```python
test_encoder_values=encoder.transform(Xtest)
```

```python
column_names=encoder.get_feature_names_out()
```

```python
column_names
```

# Create data frame of encoded Values

### Encoding Xtrain 

```python
# encoded_df=pd.DataFrame(encoder_values,columns=column_names)
# Xtrain_encoded=pd.DataFrame(train_encoder_values,columns=column_names)
Xtrain_encoded=pd.DataFrame(train_encoder_values,columns=column_names)
```

```python
Xtrain_encoded
```

### Encoding Xtest

```python
# encoded_df=pd.DataFrame(encoder_values,columns=column_names)
# Xtest_encoded=pd.DataFrame(test_encoder_values,columns=column_names)
Xtest_encoded=pd.DataFrame(test_encoder_values,columns=column_names)
```

```python
Xtest_encoded
```

# Model Creation

```python
Model=LinearRegression()
```

```python
Model.fit(Xtrain_encoded,ytrain)
```

```python
ytrainpred=Model.predict(Xtrain_encoded)
```

```python
ytestpred=Model.predict(Xtest_encoded)
```

```python
train_score=r2_score(ytrain,ytrainpred)
#ytrain is a actual train output from train data and xtrainpolypred is a predicted output from xtrainpoly which is from  train data
test_score=r2_score(ytest,ytestpred) 
#ytest is a actual test output from test data and xtestpolypred is a predicted output from xtestpoly which is from  test data
print(f'Train Score : => {train_score}\nTest Score : => {test_score}')
```

```python
train_score = Model.score(Xtrain_encoded,ytrain)
test_score = Model.score(Xtest_encoded,ytest)
print(f'Train Score : => {train_score}\nTest Score : => {test_score}')
```
