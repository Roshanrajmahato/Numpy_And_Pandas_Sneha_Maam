```python
from sklearn.datasets import load_diabetes
```

```python
df = load_diabetes(as_frame=True)
X= df.data
y= df.target
```

```python
X
```

```python
y
```

```python
from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.8,random_state=42)
```

```python
pipeline = Pipeline(
    steps=[
        ('kmeans',KMeans(n_clusters=2)),
        ('model',KNeighborsRegressor())
    ]
)

# pipeline.fit(xtrain,ytrain)
pipeline.named_steps['kmeans'].fit_transform(xtrain)
pipeline.named_steps['kmeans'].cluster_centers_
```

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer
```

```python
df = pd.read_csv("spam.csv",encoding='latin1')

X=df.v2
y=df.v1
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.8,random_state=42)
```

```python
y.value_counts()
```

```python
vectorizer = CountVectorizer(stop_words='english')

x_vector = vectorizer.fit_transform(X)
x_vector.toarray()
```

```python
sampling = RandomUnderSampler(random_state=42)


x_resample,y_resample=sampling.fit_resample(x_vector,y)
```
