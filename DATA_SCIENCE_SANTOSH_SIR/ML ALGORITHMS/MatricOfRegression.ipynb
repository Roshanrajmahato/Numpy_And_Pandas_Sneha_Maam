```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
```

```python
df = pd.read_csv(r"D:\Roshan raj Mahato\Desktop\Data Analytics\Data_Science_By_Santash_Sir\Datasets\student_performance.csv")
df.info()
```

```python
df.head()
```

```python
df['Extracurricular Activities'] = df['Extracurricular Activities'].map({'No':0,'Yes':1})
```

```python
X = df.drop('Performance Index',axis=1)
y = df['Performance Index']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
xtrain
```

```python
ytrain
```

```python
xtest
```

```python
ytest
```

```python
model = LinearRegression()
model.fit(xtrain,ytrain)
```

```python
model.coef_
# [2.85248393, 1.0169882 , 0.60861668, 0.47694148, 0.19183144] for each fearture 
# slopes are the weights
# weights :- It tell the important of features
```

```python
model.intercept_

#y = -33.92194621555611(base value of y), when all the features are zero 
```

```python
#weights ?
#it tells the important of feature
```

```python
#model.score() == r2_score()
'''
#model.score() == r2_score()

model.score(x,y)
for test => model.score(xtest,ytest)
for train => model.score(xtrain,ytrain)
'''
test_score = model.score(xtest,ytest)
train_score = model.score(xtrain,ytrain)
print(f'Train Score : => {train_score}\nTest Score : => {test_score}')
```

`Conclusion : `

`Train Data` => Model captures 98% of the pattern from train data.

`Test Data` => Model captures 98% of the pattern from test data.

The performance from both train and test score is good.

Model learns the train data pattern properly instead of memorizing and 
the performance from unseen data is also good.so this model is good for future predictions
