```python
from sklearn.svm import SVC
import pandas as pd 
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
import os
 
```

```python
df=pd.read_csv(r"D:\Roshan raj Mahato\Desktop\Data Analytics\DATA_SCIENCE_SANTOSH_SIR\Datasets\loan_approval_dataset.csv")
```

```python
df
```

```python
X = df.drop('Loan_Approved',axis=1)
y= df['Loan_Approved']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
obj_col = X.select_dtypes(include='object').columns
```

```python
num_col = X.select_dtypes(include='number').columns
```

```python
preprocessing=ColumnTransformer(
    transformers=[
        ('scalar',StandardScaler(),num_col),
        ('encode',OneHotEncoder(),obj_col) 
    ],remainder='passthrough'
)
preprocessing.fit_transform(xtrain)
```

```python
main_pipeline=Pipeline(
    steps=[
        ('pre',preprocessing),
        ('model',SVC())
    ]
)
```

```python
grid_search=GridSearchCV(
    estimator=main_pipeline,
    param_grid={
        'model__C':[0.001,0.1,11.0,100],
        'model__kernel':['linear', 'poly', 'rbf', 'sigmoid'],
        'model__gamma':[0.01,1.0,10,100],
   
    },cv=5,verbose=5,n_jobs=-1
)
grid_search.fit(xtrain,ytrain)
```
