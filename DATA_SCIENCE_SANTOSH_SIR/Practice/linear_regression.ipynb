```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r'D:\Roshan raj Mahato\Desktop\Data Analytics\Data_Science_By_Santash_Sir\Datasets\Salary_Data.csv')
```

```python
df.head()
```

```python
X = df[['YearsExperience']]
```

```python
y = df['Salary']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression() # creating model or object
```

```python
# .fit Used to find the pattern and based on that its going to train it (.fit is the method of object(model))
model.fit(X,y)
```

```python
model.intercept_ # .intercept_ slope :- the base or initail of salary is 25792.200198668717 if experience is 0.
```

```python
model.coef_ # .coef_ :- it slope (Rate of change of salary is  with respect to y(experience) )
```

```python
# .predict(X) :- To get the predicted value from model 
#  Internal 
#  y = mX + C
#  y=9449.96232146 X + 25792.200198668717
yprict=model.predict(X)
```

```python
df['Predicted_Values']=yprict
```

```python
df['Error']=df['Salary']-df['Predicted_Values']
```

```python
import matplotlib.pyplot as plot
```

```python
# With Actual values
plot.scatter(X,y)
# With Predicted Values
plot.scatter(X,yprict)
```

```python
from sklearn.model_selection import train_test_split
'''
syntax:-
xtrain,xtest,ytrain,ytest = 
train_test_split(x,y,random_state=42,train_size=0.8)'''
```
