# PROBLEM PYTHON 1: Find Pairs with Given Target Sum

**Problem Statement**

Write a Python program to find all pairs of numbers in a list whose **sum equals a given target value**.

**Input**  
`l1 = [2, 3, 1, 4, 2, 5, 6, 7, 1, 2, 4, 8, 1, 4, 5]`  
`target = 6`

**Output**  
`[(2,4), (3,3), (1,5), ...]` (pairs whose sum equals target)

## Best Optimized Approach (Using Hash Set — Recommended)

````python
l1 = [2, 3, 1, 4, 2, 5, 6, 7, 1, 2, 4, 8, 1, 4, 5]
target = int(input("Enter a value: "))
seen = set()
res = []
for num in l1:
    complement = target - num
    if complement in seen:
        res.append((complement, num))
    seen.add(num)
print(res)
````

# Problem: Replace Elements by Its Rank in the Array

## 📝 Problem Statement
You are given an array of **N** integers. Your task is to replace each element of the array with its **rank**.

**Rules for Ranking:**
1. The rank of an element is its position when the array is sorted in ascending order.
2. The smallest element gets a rank of 1.
3. If two elements are equal, they must be assigned the **same rank**.
4. The next rank assigned after a tie should be the immediate next integer (e.g., if two elements share rank 3, the next unique element gets rank 4).

---

### Example 1
**Input:**
```text
6
20 15 26 2 98 6
```
**Explanation:**
1. Sorted array: `2, 6, 15, 20, 26, 98`
2. Assign Ranks: `2→1, 6→2, 15→3, 20→4, 26→5, 98→6`
3. Map back to original order: `20(4), 15(3), 26(5), 2(1), 98(6), 6(2)`

**Output:**
```text
4 3 5 1 6 2
```
### Example 2 (With Duplicates)
**Input:**
```text
7
1 5 8 15 8 25 9
```
**Explanation:**
1. Sorted Unique: `1, 5, 8, 9, 15, 25`
2. Assign Ranks: `1→1, 5→2, 8→3, 9→4, 15→5, 25→6`
3. Map back to original: `1(1), 5(2), 8(3), 15(5), 8(3), 25(6), 9(4)`

**Output:**
```text
1 2 3 5 3 6 4
```


````python
def replaceWithRank(arr):
    # Step 1: Initialize an empty map and starting rank
    rank_map = {}
    rank = 1
    
    # Step 2: Sort the array to determine positions
    # sorted(arr) handles the ascending order logic
    for num in sorted(arr):
        # Step 3: Assign rank only if the number is seen for the first time
        # This handles the 'same rank for equal elements' rule
        if num not in rank_map:
            rank_map[num] = rank
            rank += 1
            
    # Step 4: Use list comprehension to replace original elements with ranks
    return [rank_map[num] for num in arr]

# --- Input Section ---
try:
    n = int(input().strip())
    # Read the second line and convert space-separated strings to integers
    arr = list(map(int, input().split()))

    result = replaceWithRank(arr)

    # Step 5: Print result using the unpacking operator (*) to space-separate output
    print(*result)
except EOFError:
    pass
````
