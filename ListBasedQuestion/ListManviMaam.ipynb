# Python Problem: Linear Search in Array


## Problem Statement

Write a Python program to **search an element in an array** using **Linear Search**.

---

**Input**<br>
`arr = [10, 5, 3, 6, 2, 8, 4]`
`x = 3`

**Output**<br>
`Found 3 at index 2`

**Explanation**

Linear Search checks each element one by one until the target element is found.

- Compare each element with `x`
- If found, store index
- Print index of found element

**Algorithm Steps**

- Start
- Define array
- Define search element
- Initialize `ans = -1`
- Traverse array
- If element matches, store index
- After loop check result
- Print found message
- Stop


````python
def searchInArray():
    arr = [10, 5, 3, 6, 2, 8, 4]
    x = 3
    ans = -1
    for i in range(len(arr)):
        if arr[i] == x:
            ans = i
    if ans == -1:
        print("Not Found")
    else:
        print("Found", x, "at index", ans)


searchInArray()
````

---

**Input**

# Python Problem: Check if Array is Sorted

**Problem Statement**

Write a Python program to **check whether an array is sorted in ascending order**.

---

**Input**

`arr = [1, 2, 3, 4, 5]`

**Output**

`True`

**Explanation**

- Compare each element with the previous element.
- If any element is smaller than the previous one, the array is not sorted.
- Otherwise, the array is sorted.

---

**Algorithm Steps**

- Start
- Define function to check sorted array
- Traverse array from index 1
- Compare current element with previous element
- If smaller, return False
- Otherwise continue
- Return True
- Stop



### 1️⃣ Your Code — Proper Clean Version


````python
def isSorted(arr):
    check = True
    for i in range(1, len(arr)):
        if arr[i] < arr[i - 1]:
            check = False
            break
    return check


arr = [1, 2, 3, 4, 5]
result = isSorted(arr)
print("Is Sorted:", result)
````


### 2️⃣ Optimized Version (Return Immediately) ⭐ Better


````python
def isSorted(arr):

    for i in range(1, len(arr)):
        if arr[i] < arr[i - 1]:
            return False

    return True


arr = [1, 2, 3, 4, 5]

print("Is Sorted:", isSorted(arr))
````


### 3️⃣ More Pythonic Version (Using `all()`)

````python
def isSorted(arr):
    return all(arr[i] >= arr[i - 1] for i in range(1, len(arr)))


arr = [1, 2, 3, 4, 5]

print("Is Sorted:", isSorted(arr))
````


### 4️⃣ Reverse Sorted Check (Extra Useful)

````python
def isSorted(arr):

    ascending = True
    descending = True

    for i in range(1, len(arr)):
        if arr[i] < arr[i - 1]:
            ascending = False

        if arr[i] > arr[i - 1]:
            descending = False

    return ascending or descending


arr = [5, 4, 3, 2, 1]

print("Is Sorted:", isSorted(arr))
````


### 5️⃣ Best Practice Version (Handles Empty List)


````python
def isSorted(arr):

    if len(arr) <= 1:
        return True

    for i in range(1, len(arr)):
        if arr[i] < arr[i - 1]:
            return False

    return True


arr = [1, 2, 3, 4, 5]

print("Is Sorted:", isSorted(arr))
````

**Algorithm Steps**

- Traverse array

- Return index

### 1️⃣ Your Current Code — Clean Version (Baseline)

````python
def firstOccurrence(arr, x):

    for i in range(len(arr)):
        if arr[i] == x:
            return i

    return -1


def lastOccurrence(arr, x):

    lastIndex = -1

    for i in range(len(arr)):
        if arr[i] == x:
            lastIndex = i

    return lastIndex


arr = [1, 2, 3, 2, 5, 4, 2, 8, 9]
x = 2

print("First Occurrence Index:", firstOccurrence(arr, x))
print("Last Occurrence Index:", lastOccurrence(arr, x))
````

````python
def firstOccurrence(arr, x):

    for i in range(len(arr)):
        if arr[i] == x:
            return i

    return -1


def lastOccurrence(arr, x):

    for i in range(len(arr) - 1, -1, -1):
        if arr[i] == x:
            return i

    return -1


arr = [1, 2, 3, 2, 5, 4, 2, 8, 9]
x = 2

print("First Occurrence Index:", firstOccurrence(arr, x))
print("Last Occurrence Index:", lastOccurrence(arr, x))
````

### 4️⃣ Built-in Optimized Version ⭐ Shortest

````python
def firstOccurrence(arr, x):

    if x in arr:
        return arr.index(x)

    return -1


def lastOccurrence(arr, x):

    if x in arr:
        return len(arr) - 1 - arr[::-1].index(x)

    return -1


arr = [1, 2, 3, 2, 5, 4, 2, 8, 9]
x = 2

print("First Occurrence Index:", firstOccurrence(arr, x))
print("Last Occurrence Index:", lastOccurrence(arr, x))
````

### 5️⃣ Most Optimized — Binary Search (Sorted Array Only) ⭐⭐⭐

````python
def firstOccurrence(arr, x):

    low = 0
    high = len(arr) - 1
    ans = -1

    while low <= high:

        mid = (low + high) // 2

        if arr[mid] == x:
            ans = mid
            high = mid - 1

        elif arr[mid] < x:
            low = mid + 1

        else:
            high = mid - 1

    return ans


def lastOccurrence(arr, x):

    low = 0
    high = len(arr) - 1
    ans = -1

    while low <= high:

        mid = (low + high) // 2

        if arr[mid] == x:
            ans = mid
            low = mid + 1

        elif arr[mid] < x:
            low = mid + 1

        else:
            high = mid - 1

    return ans


arr = [1, 2, 2, 2, 3, 4, 5]
x = 2

print("First Occurrence Index:", firstOccurrence(arr, x))
print("Last Occurrence Index:", lastOccurrence(arr, x))
````

### 6️⃣ Find Both First & Last in One Function ⭐ Useful

````python
def firstLastOccurrence(arr, x):

    first = -1
    last = -1

    for i in range(len(arr)):

        if arr[i] == x:

            if first == -1:
                first = i

            last = i

    return first, last


arr = [1, 2, 3, 2, 5, 4, 2, 8, 9]
x = 2

first, last = firstLastOccurrence(arr, x)

print("First Occurrence Index:", first)
print("Last Occurrence Index:", last)
````

# Python Problem: Count Occurrences of Element in Array

Write a Python program to **count how many times an element appears** in an array.

## Algorithm Steps

### 1️⃣ Your Code — Clean Version

````python
def countOccurrences(arr, x):
    count = 0
    for i in range(len(arr)):
        if arr[i] == x:
            count += 1
    return count

# Example execution
arr = [1, 2, 3, 2, 5]
x = 2

print("Count:", countOccurrences(arr, x))
````

````python
def countOccurrences(arr, x):
    count = 0
    for num in arr:
        if num == x:
            count += 1
    return count

arr = [1, 2, 3, 2, 5]
x = 2

print("Count:", countOccurrences(arr, x))
````


### 4️⃣ Built-in Optimized Version ⭐⭐ Best Short Code

### 5️⃣ Most Optimized (Sorted Array — Binary Search) ⭐⭐⭐

````python
def firstOccurrence(arr, x):

    low = 0
    high = len(arr) - 1
    ans = -1

    while low <= high:

        mid = (low + high) // 2

        if arr[mid] == x:
            ans = mid
            high = mid - 1

        elif arr[mid] < x:
            low = mid + 1

        else:
            high = mid - 1

    return ans


def lastOccurrence(arr, x):

    low = 0
    high = len(arr) - 1
    ans = -1

    while low <= high:

        mid = (low + high) // 2

        if arr[mid] == x:
            ans = mid
            low = mid + 1

        elif arr[mid] < x:
            low = mid + 1

        else:
            high = mid - 1

    return ans


def countOccurrences(arr, x):

    first = firstOccurrence(arr, x)

    if first == -1:
        return 0

    last = lastOccurrence(arr, x)

    return last - first + 1


arr = [1, 2, 2, 2, 3, 4, 5]
x = 2

print("Count:", countOccurrences(arr, x))
````

### 6️⃣ Find First, Last & Count Together ⭐ Very Useful

````python
def occurrenceDetails(arr, x):
    first = -1
    last = -1
    count = 0

    for i in range(len(arr)):
        if arr[i] == x:
            if first == -1:
                first = i
            last = i
            count += 1
    return first, last, count


arr = [1, 2, 3, 2, 5]
x = 2

first, last, count = occurrenceDetails(arr, x)
print("First Index:", first)
print("Last Index:", last)
print("Count:", count)
````

# Python Problem: Find Unique Element in Array


**Problem Statement**

Write a Python program to **find the unique element** in an array where all other elements appear **twice**.

The program should:

- Compare each element with every other element
- Mark duplicate elements as `-1`
- Find the element that appears only once
- Return the unique element

---

**Input**

`n = 5`
`arr = [1, 2, 3, 2, 1]`

**Output**

`Unique element: 3`

**Explanation**

- Compare each element with the remaining elements.
- If two elements are equal, mark both as `-1`.
- After marking duplicates, only the unique element remains positive.
- Traverse the array again to find the element greater than `0`.

Example:
`[1, 2, 3, 2, 1] → [-1, -1, 3, -1, -1] → Unique = 3`

---

**Algorithm Steps**

- Start
- Read size of array
- Create array
- Read array elements
- Traverse array using nested loops
- Compare elements
- If duplicate found, mark both as `-1`
- Traverse array again
- Find element greater than `0`
- Store it as unique element
- Print unique element
- Stop

### 1️⃣ Your Code — Cleaned Version (Same Logic)

````python
def findUnique(arr):
    n = len(arr)

    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] == arr[j]:
                arr[i] = -1
                arr[j] = -1
    ans = -1

    for i in range(n):
        if arr[i] > 0:
            ans = arr[i]
            
    return ans

n = int(input("Enter array size: "))
arr = []
print("Enter", n, "elements")

for i in range(n):
    arr.append(int(input()))
print("Unique element:", findUnique(arr))
````

### 2️⃣ Better Nested Loop (Without Modifying Array)

````python
def findUnique(arr):

    n = len(arr)
    for i in range(n):
        count = 0
        for j in range(n):
            if arr[i] == arr[j]:
                count += 1
        if count == 1:
            return arr[i]
    return -1

n = int(input("Enter array size: "))
arr = []
print("Enter", n, "elements")

for i in range(n):
    arr.append(int(input()))
print("Unique element:", findUnique(arr))
````

### 3️⃣ Optimized Using Dictionary ⭐ Best Interview Method


````python
def findUnique(arr):

    freq = {}

    for num in arr:
        freq[num] = freq.get(num, 0) + 1

    for num in arr:
        if freq[num] == 1:
            return num

    return -1


n = int(input("Enter array size: "))
arr = []
print("Enter", n, "elements")

for i in range(n):
    arr.append(int(input()))
    
print("Unique element:", findUnique(arr))
````

### 4️⃣ Using `Counter` ⭐⭐ Very Clean

````python
from collections import Counter

def findUnique(arr):
    freq = Counter(arr)
    for num in arr:
        if freq[num] == 1:
            return num
    return -1
````

### 5️⃣ Most Optimized — XOR Method ⭐⭐⭐ (Special Case)

````python
def findUnique(arr):

    result = 0

    for num in arr:
        result ^= num

    return result


n = int(input("Enter array size: "))
arr = []
print("Enter", n, "elements")

for i in range(n):
    arr.append(int(input()))

print("Unique element:", findUnique(arr))
````

### 6️⃣ Find All Unique Elements (Not Just One)

````python
def findAllUnique(arr):
    freq = {}

    for num in arr:
        freq[num] = freq.get(num, 0) + 1
    unique = []

    for num in arr:
        if freq[num] == 1:
            unique.append(num)
    return unique


arr = [1, 2, 3, 2, 4, 5, 4]
print("Unique elements:", findAllUnique(arr))
````

# Python Problem: Count Pair Sum Equal to Target

**Problem Statement**

Write a Python program to **count the number of pairs** in an array whose **sum is equal to a given target value**.

The program should:

- Compare every pair of elements
- Check whether their sum equals the target
- Count all such valid pairs
- Return the total number of pairs

---

**Input**

`n = 6`
`arr = [1, 2, 3, 4, 5, 6]`
`target = 7`

**Output**

`3`

**Explanation**

Check all possible pairs:

- `(1 + 6) = 7` ✔
- `(2 + 5) = 7` ✔
- `(3 + 4) = 7` ✔

Total valid pairs = **3**

---

**Algorithm Steps**

- Start
- Read size of array
- Create array
- Read array elements
- Read target value
- Initialize pair count as `0`
- Traverse array using nested loops
- For each pair, check if sum equals target
- If equal, increase pair count
- After checking all pairs, return count
- Print result
- Stop



### 1️⃣ Your Code — Clean Version (Baseline)

````python
def tripletSum(arr, target):

    arr.sort()

    n = len(arr)

    ans = 0

    for i in range(n - 2):

        # Skip duplicate first elements
        if i > 0 and arr[i] == arr[i - 1]:
            continue

        left = i + 1
        right = n - 1

        while left < right:

            s = arr[i] + arr[left] + arr[right]

            if s == target:

                ans += 1

                # Skip duplicates
                while left < right and arr[left] == arr[left + 1]:
                    left += 1

                while left < right and arr[right] == arr[right - 1]:
                    right -= 1

                left += 1
                right -= 1

            elif s < target:

                left += 1

            else:

                right -= 1

    return ans
````

Steps:

---

**Algorithm Steps**

### 1️⃣ Your Code — Two Pointer Method (Already Optimized ⭐)

# 1️⃣ Subarray Sum Equals K


**Given:**

* An array of integers
* A number **k**

**Task:**

Find the **number of subarrays** whose **sum equals k**.

---

**🎯 Example**

```text
arr = [1, 2, 3]
k = 3
```

Valid subarrays:

```text
[1,2] → sum = 3
[3]   → sum = 3
```

Output:

```text
2
```



### 1️⃣ Brute Force Approach (Understand First)


````python
def subarraySumBrute(arr, k):

    n = len(arr)
    count = 0

    for i in range(n):

        total = 0

        for j in range(i, n):

            total += arr[j]

            if total == k:
                count += 1

    return count


# Example
arr = [1, 2, 3]
k = 3

print(subarraySumBrute(arr, k))
````


### 2️⃣ Optimized Approach — Prefix Sum + HashMap ⭐⭐⭐


````python
def subarraySum(arr, k):

    prefix_sum = 0

    count = 0

    prefix_map = {}

    # Base case
    prefix_map[0] = 1

    for num in arr:

        prefix_sum += num

        # Check if needed sum exists
        if prefix_sum - k in prefix_map:
            count += prefix_map[prefix_sum - k]

        # Store prefix sum
        if prefix_sum in prefix_map:
            prefix_map[prefix_sum] += 1
        else:
            prefix_map[prefix_sum] = 1

    return count


# Example
arr = [1, 2, 3]
k = 3

print(subarraySum(arr, k))
````

### ⭐ Cleaner Python Version (Using defaultdict)

````python
from collections import defaultdict

def subarraySum(arr, k):

    prefix_sum = 0
    count = 0

    prefix_map = defaultdict(int)

    prefix_map[0] = 1

    for num in arr:

        prefix_sum += num

        count += prefix_map[prefix_sum - k]

        prefix_map[prefix_sum] += 1

    return count
````


# 3️⃣ Longest Subarray Sum = K ⭐⭐⭐⭐

**🧠 Problem**

Find **length of longest subarray** with sum = **k**

---

**🎯 Example**

```text
arr = [1, -1, 5, -2, 3]
k = 3
```

Valid:

```text
[1, -1, 5, -2] → sum = 3
```

Output:

```text
4
```


### ✅ Best Code

````python
def longestSubarraySumK(arr, k):

    prefix_sum = 0

    max_length = 0

    prefix_map = {}

    for i in range(len(arr)):

        prefix_sum += arr[i]

        if prefix_sum == k:
            max_length = i + 1

        if prefix_sum - k in prefix_map:
            length = i - prefix_map[prefix_sum - k]
            max_length = max(max_length, length)

        if prefix_sum not in prefix_map:
            prefix_map[prefix_sum] = i

    return max_length


# Example
arr = [1, -1, 5, -2, 3]
k = 3

print(longestSubarraySumK(arr, k))
````


# 4️⃣ Subarray Sum Divisible by K ⭐⭐⭐⭐⭐



This is **advanced** but extremely popular.

---

**🧠 Problem**

Count subarrays whose sum is **divisible by k**

---

**🎯 Example**

```text
arr = [4, 5, 0, -2, -3, 1]
k = 5
```

Output:

```text
7
```

### ✅ Best Code

````python
from collections import defaultdict

def subarrayDivisibleByK(arr, k):

    prefix_sum = 0

    count = 0

    remainder_map = defaultdict(int)

    remainder_map[0] = 1

    for num in arr:

        prefix_sum += num

        remainder = prefix_sum % k

        if remainder < 0:
            remainder += k

        count += remainder_map[remainder]

        remainder_map[remainder] += 1

    return count


# Example
arr = [4, 5, 0, -2, -3, 1]
k = 5

print(subarrayDivisibleByK(arr, k))
````
