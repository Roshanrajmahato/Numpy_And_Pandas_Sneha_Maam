## Problem Statement 1: Count Frequency of Elements in a List

Write a Python program to count how many times each element appears in a given list using a dictionary.

**Input:**
`l1 = [1,2,3,2,1,3,4,6,5,3,2,3,1,3,2,5,8,7,5,6,7]`

**Output:**
`{1: 3, 2: 4, 3: 5, 4: 1, 6: 2, 5: 3, 8: 1, 7: 2}`

**Algorithm Steps :-**
- Start
- Create list `l1`
- Create empty dictionary `d1`
- Traverse each element `i` in `l1`
- If `i` not in `d1`, set `d1[i] = 1`
- Else increase `d1[i]` by `1`
- Print `d1`
- Stop

```python
l1=[1,2,3,2,1,3,4,6,5,3,2,3,1,3,2,5,8,7,5,6,7]
d1={}
for i in l1:
    if i not in d1:
        d1[i]=1
    else:
        d1[i]+=1
d1
```

# Python Problem 2: Flatten a Nested List

## Problem Statement
Write a Python program to flatten a list that contains integers and sublists.If an element is a list, extract its elements and add them to a new list.If the element is not a list, add it directly to the new list.

**Input:**
`l1 = [1,2,[34,5,6],6,[7,1],6,9,1,[4,2,7,9]]`

**Output:**
`[1,2,34,5,6,6,7,1,6,9,1,4,2,7,9]`

**Algorithm Steps**
- Start
- Create list `l1`
- Create empty list `new`
- Traverse each element `i` in `l1`
- Check if `i` is a list using `isinstance(i, list)`
- If true, traverse each element `j` inside that list
- Append `j` to list `new`
- Else append `i` directly to list `new`
- Continue until all elements are processed
- Print list `new`
- Stop

```python
l1=[1,2,[34,5,6],6,[7,1],6,9,1,[4,2,7,9]]
new=[]
for i in l1:
    if isinstance(i,list):
        for j in i:
            new+=[j]
    else:
        new+=[i]

print(new)
```
