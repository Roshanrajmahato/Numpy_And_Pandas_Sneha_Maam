# Problem 1: The Smart Elevator

A company has installed a smart elevator in a 10-floor building (floors numbered 0 to 9).
The elevator starts at floor 0.
You are given a sequence of floor requests.

**Rules**
* If the requested floor is higher than current floor $\rightarrow$ elevator moves up.
* If lower $\rightarrow$ elevator moves down.
* Moving 1 floor takes **2 seconds**.
* Stopping at a floor takes **1 second**.
* If the request is the same as current floor $\rightarrow$ only **1 second** (stop time).

**Your task**
Calculate the total time taken to serve all requests in order.

**Input**<br>
$n$  
$f_1, f_2, f_3, \dots, f_n$

**Constraints**
* $1 \le n \le 100$
* $0 \le f_i \le 9$

```python
# METHOD 1
n = int(input())
current_floor = 0
total_time = 0

for _ in range(n):
    req = int(input())
    total_time += abs(req - current_floor) * 2
    total_time += 1
    current_floor = req

print(total_time)
```

```python
# METHOD 2
def calculate_total_time(requests):
    current_floor = 0
    total_time = 0
    for floor in requests:
        if floor == current_floor:
            total_time += 1
        else:
            move = abs(floor - current_floor)
            total_time += move * 2
            total_time += 1
            current_floor = floor
    return total_time

n = int(input())
requests = []
for _ in range(n):
    req = int(input())
    requests+=[req]
print(calculate_total_time(requests))  # Output: 13
```

```python
# METHOD 3
n = int(input())
requests = list(map(int, input().split()))
current_floor = 0
total_time = 0 # For Storing Continous Result From Some Operations # Continous Updation It Will Have 
for req in requests:
    total_time += abs(req - current_floor) * 2 # If i have to do Operation On Values and the same operation i have to perform for different values also  
    total_time += 1                           # Then A Result Of that Operation we get ,If we have to Store Or Save it Continously , Then Used one global Variable like this. 
    current_floor = req
print(total_time)
```

# Problem 2: The Digital Locker System

A school locker has a **4-digit PIN**.

A student enters a number.
If the number:
* Contains exactly **2 even digits**
* The sum of digits is **divisible by 3**
* **No digit repeats**

Then locker opens. Otherwise, it remains locked.

**Input**:<br>
A 4-digit integer

**Output**:<br>
Print **OPEN** or **LOCKED**

---

```python
def getNoOfDigits(n):
    digits = 0
    if n == 0: return 1 # Basic safety for 0
    temp = n
    while temp > 0:
        digits += 1
        temp //= 10
    return digits

def getSumOfDigits(n):
    sum = 0
    temp = n
    while temp > 0:
        sum += temp % 10
        temp //= 10
    return sum

def getEvenDigits(n):
    evenDigits = 0
    temp = n
    while temp > 0:
        rem = temp % 10
        if rem % 2 == 0:
            evenDigits += 1
        temp //= 10
    return evenDigits

def check(n):
    originalDigits = getNoOfDigits(n)
    st = set()
    temp = n
    while temp > 0:
        rem = temp % 10
        st.add(rem)
        temp //= 10
    
    if len(st) == originalDigits: 
        return True
    else:
        return False


if __name__ == "__main__":
    n = int(input())
    # Exact logic from the image's if-statement
    if (check(n) == False or getNoOfDigits(n) != 4 or getSumOfDigits(n) % 3 != 0 or getEvenDigits(n) != 2):
        print("LOCKED") 
    else:
        print("OPEN")
```

# Problem 3: The Warehouse Box Arrangement

A warehouse receives **N** boxes, each having a weight.

Manager wants to form pairs such that:
* Each pair must have **equal total weight**.
* Each box can be used **only once**.
* If impossible $\rightarrow$ print **-1**.

Otherwise print:
* **Number of pairs formed**.

### Input:
$N$  
$w_1, w_2, w_3, \dots, w_N$

### Constraints:
* $1 \le N \le 10^5$
* $1 \le w_i \le 10^4$

---

Would you like me to explain the logic for sorting and pairing the boxes to solve this?

```python
n = int(input())
if n % 2 != 0:
    print("-1")
    exit()
v = list(map(int, input().split()))
v.sort()
target = v[0] + v[len(v) - 1]
pairs = 0
i = 0
j = len(v) - 1
while i < j:
    if v[i] + v[j] != target:
        print("-1")
        exit()   
    pairs += 1 
    i += 1
    j -= 1

print(pairs)
```

## Problem 5: The Message Decoder

You receive a coded message as a string.

### Rule:
* Each digit represents how many times the next character should be repeated.
* If no digit before character $\rightarrow$ print it once.

### Input:
Example: **3a2b1c**

### Output:
**aaabbc**

---

Would you like me to show you how to handle the "no digit before character" rule in Python using a loop?

```python
s = input()
i = 0
while i < len(s):
    if s[i].isdigit():
        count = int(s[i])  # int count = s[i] - '0';
        print(s[i + 1] * count, end="")       
        i += 1
    else:
        print(s[i], end="")   
    i += 1
print()
```

### 2. Equilibrium Index of an Array
Given an array, find an index such that the sum of elements on its left is equal to the sum of elements on its right. Return -1 if no such index exists.

---

```python
# 1. Get the list of numbers
nums = [1, 7, 3, 6, 5, 6]
n = len(nums)

# 2. Calculate the Total Sum (ts)
total_sum = sum(nums)
print(f"Total Sum: {total_sum}")

# 3. Keep track of the Left Sum (ls)
left_sum = 0

# 4. Walk through the list one by one
for i in range(n):
    # Calculate the Right Sum (rs)
    # Right Sum = Total - Left - Current Number
    right_sum = total_sum - left_sum - nums[i]
    
    # 5. Check if they balance
    if left_sum == right_sum:
        print(f"Pivot found at index: {i}")
        exit() # Stop the program
        
    # 6. If not balanced, add the current number to the Left Sum and move on
    left_sum += nums[i]

# 7. If we finish the loop and find nothing
print("-1")
```

### 3. First and Last Occurrence of an Element
Given a sorted array and a target value, find the first and last index of the target using binary search techniques. If not found, return -1 -1.

---

```python
def getFirst(v, target):
    n = len(v)
    lo = 0
    hi = n - 1
    ans = -1
    
    while lo <= hi:
        # mid = lo + (hi - lo) / 2
        mid = lo + (hi - lo) // 2
        
        if v[mid] == target:
            ans = mid
            hi = mid - 1  # Look on the left side for an earlier occurrence
        elif v[mid] < target:
            lo = mid + 1
        else:
            hi = mid - 1
            
    return ans

def getSecond(v, target):
    n = len(v)
    lo = 0
    hi = n - 1
    ans = -1
    
    while lo <= hi:
        mid = lo + (hi - lo) // 2
        
        if v[mid] == target:
            ans = mid
            lo = mid + 1  # Look on the right side for a later occurrence
        elif v[mid] < target:
            lo = mid + 1
        else:
            hi = mid - 1
            
    return ans


n = int(input("Enter n: "))

# vector<int> v(n)
v = []
print(f"Enter {n} sorted numbers:")
for _ in range(n):
    v.append(int(input()))
    
# cin >> target
target = int(input("Enter target: "))

first = getFirst(v, target)
second = getSecond(v, target)

# cout << "First : " << first << " Second : " << second;
print(f"First : {first} Second : {second}")
        


```

### 4. Majority Element
Given an array of size N, find the element that appears more than N/2 times. If no such element exists, return -1.

---

```python
def main():
    # cin >> n;
    n = int(input())

    # vector<int> v(n);
    v = []
    # for(int i = 0; i < n; i++){ cin >> v[i]; }
    # (Reading inputs until we have n elements)
    while len(v) < n:
        line = input().split()
        for x in line:
            v.append(int(x))

    # map<int, int> mp;
    mp = {}

    # for(auto it : v){ mp[it]++; }
    for val in v:
        mp[val] = mp.get(val, 0) + 1

    # for(auto it : mp){ ... }
    # (Note: sorted(mp.keys()) ensures alphabetical/numerical order 
    # to match C++ std::map behavior)
    for key in sorted(mp.keys()):
        count = mp[key]
        
        # if(it.second >= n / 2){ ... }
        if count >= (n // 2):
            # cout << it.first;
            print(key)
            # return 0;
            return

    # cout << -1 << endl;
    print("-1")

if __name__ == "__main__":
    main()
```

### 5. Two Sum (Sorted Input)
Given a sorted array and a target sum, find two numbers such that they add up to the target. Return their indices or values.

---

```python
# --- Algorithmic Steps ---
# 1. Take the size of the array (n) as input.
# 2. Loop 'n' times to take each array element as a separate input.
# 3. Take the target sum as the final input.
# 4. Initialize two pointers: 'i' at the start (0) and 'j' at the end (n-1).
# 5. While pointers haven't met:
#    - If v[i] + v[j] == target: success, print indices and exit.
#    - If v[i] + v[j] < target: sum is too small, move left pointer (i) right.
#    - If v[i] + v[j] > target: sum is too large, move right pointer (j) left.
# 6. If loop ends without a match, print -1.

n = int(input("Enter n: "))

# Create the list by taking input n times
v = []
print(f"Enter {n} sorted numbers:")
for _ in range(n):
    v.append(int(input()))

# Take the target sum input
target = int(input("Enter target: "))

# Initialize pointers
i = 0
j = n - 1

# Two-pointer logic
found = False
while i < j:
    current_sum = v[i] + v[j]
    # If the sum matches the target, print indices and exit
    if current_sum == target:
        print(f"{i} {j}")
        found = True
        break
    # Sum is too low, move the left pointer to a larger value
    elif current_sum < target:
        i += 1
    # Sum is too high, move the right pointer to a smaller value
    else:
        j -= 1

# If no pair was found after the loop finishes
if not found:
    print("-1")
```

### 6. Transpose of a Matrix
Given an N x M matrix, compute its transpose (rows become columns and columns become rows).

---

```python
# --- Algorithmic Steps ---
# 1. Take row (r) and column (c) counts as input.
# 2. Use nested loops to take input for each cell v[i][j].
# 3. Print the original matrix row by row.
# 4. Transpose logic: Swap the loop order. 
#    - Outer loop runs through columns (j)
#    - Inner loop runs through rows (i)
# 5. Print v[i][j] in this new order to show the transpose.

# cin >> r >> c;
r = int(input("Enter number of rows (r): "))
c = int(input("Enter number of columns (c): "))

# vector<vector<int>> v(r, vector<int>(c, 0));
v = [[0 for _ in range(c)] for _ in range(r)]

# Nested loops for input
for i in range(r):
    for j in range(c):
        v[i][j] = int(input(f"Enter element for v[{i}][{j}]: "))

# Print the original matrix
for i in range(r):
    for j in range(c):
        print(v[i][j], end=" ")
    print() # cout << endl;

print("The transpose of the matrix is : ")

# Transpose printing logic (j then i)
for j in range(c):
    for i in range(r):
        # Accessing by [i][j] but the loops are swapped
        print(v[i][j], end=" ")
    print() # cout << endl;
```
