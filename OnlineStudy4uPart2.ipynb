# Problem: Replace Elements by Its Rank in the Array

## 📝 Problem Statement
You are given an array of **N** integers. Your task is to replace each element of the array with its **rank**.

**Rules for Ranking:**
1. The rank of an element is its position when the array is sorted in ascending order.
2. The smallest element gets a rank of 1.
3. If two elements are equal, they must be assigned the **same rank**.
4. The next rank assigned after a tie should be the immediate next integer (e.g., if two elements share rank 3, the next unique element gets rank 4).

---

### Example 1
**Input:**
```text
6
20 15 26 2 98 6
```
**Explanation:**
1. Sorted array: `2, 6, 15, 20, 26, 98`
2. Assign Ranks: `2→1, 6→2, 15→3, 20→4, 26→5, 98→6`
3. Map back to original order: `20(4), 15(3), 26(5), 2(1), 98(6), 6(2)`

**Output:**
```text
4 3 5 1 6 2
```
### Example 2 (With Duplicates)
**Input:**
```text
7
1 5 8 15 8 25 9
```
**Explanation:**
1. Sorted Unique: `1, 5, 8, 9, 15, 25`
2. Assign Ranks: `1→1, 5→2, 8→3, 9→4, 15→5, 25→6`
3. Map back to original: `1(1), 5(2), 8(3), 15(5), 8(3), 25(6), 9(4)`

**Output:**
```text
1 2 3 5 3 6 4
```

---

Would you like me to explain the **Big O complexity** of this solution or show how to solve it without using the `sorted()` function?

## 🛣️ Algorithmic Steps

To solve this efficiently ($O(N \log N)$ time complexity), follow these steps:

1.  **Input:** Read the size of the array $N$ and the array elements.
2.  **Sort & Unique:** Create a sorted version of the unique elements from the original array.
3.  **Map Creation:** Use a dictionary (hash map) to store the rank of each unique element.
    * Iterate through the sorted array.
    * If an element is not already in your map, assign it the current `rank` value and increment the `rank` by 1.
4.  **Transformation:** Iterate through the *original* array and replace each number with its corresponding rank from the map.
5.  **Output:** Print the resulting ranks separated by spaces.


````python
def replaceWithRank(arr):
    # Step 1: Initialize an empty map and starting rank
    rank_map = {}
    rank = 1
    
    # Step 2: Sort the array to determine positions
    # sorted(arr) handles the ascending order logic
    for num in sorted(arr):
        # Step 3: Assign rank only if the number is seen for the first time
        # This handles the 'same rank for equal elements' rule
        if num not in rank_map:
            rank_map[num] = rank
            rank += 1
            
    # Step 4: Use list comprehension to replace original elements with ranks
    return [rank_map[num] for num in arr]

# --- Input Section ---
try:
    n = int(input().strip())
    # Read the second line and convert space-separated strings to integers
    arr = list(map(int, input().split()))

    result = replaceWithRank(arr)

    # Step 5: Print result using the unpacking operator (*) to space-separate output
    print(*result)
except EOFError:
    pass
````

# Remove Characters from First String Present in Second String

**Problem Statement:**
You are given two strings **str1** and **str2**.
Your task is to remove all characters from **str1** that are present in **str2**.
Return the modified string after removing the common characters.
If no characters from **str1** are present in **str2**, return **str1** as it is.

**Input Format:**
* The first line contains a string **str1**.
* The second line contains a string **str2**.

**Output Format:**
* Print the modified string after removing all characters from **str1** that are present in **str2**.

**Constraints:**
* $1 \leq \text{length of str1} \leq 10^4$
* $1 \leq \text{length of str2} \leq 10^4$
* Strings contain only lowercase English alphabets.

---

***Sample Case 1:***
**Input:**
```text
abcdef
cefz
```
**Output:**
```text
abd
```

***Sample Case 2:***
**Input:**
```text
xyzpw
lmno
```
**Output:**
```text
xyzpw
```

***Sample Case 3:
**Input:**
```text
aaaaaa
a
```
**Output:**
*(Empty string)*

### Code To Write

````python
def removeChars(str1, str2):
    # Create a set for fast lookup
    remove_set = set(str2)
    result = []
    
    for ch in str1:
        if ch not in remove_set:
            result.append(ch)
            
    return "".join(result)

str1 = input().strip()
str2 = input().strip()
print(removeChars(str1, str2))
````


### ✅ Correct Logic (Simple Way)


### 🚀 Better Optimized Version (Using `set`)


````python
str1 = "roshan"
str2 = "har"

remove_set = set(str2)

rm = ""

for ch in str1:
    if ch not in remove_set:
        rm += ch

print(rm)
````


### 🎯 One-Line Pythonic Way


````python
str1 = "roshan"
str2 = "har"

rm = "".join(ch for ch in str1 if ch not in str2)

print(rm)
````

## BY Two `for` Loop


### ✅ Correct Two `for` Loop Method


````python
str1 = "roshan"
str2 = "har"

rm = ""

for i in str1:          # Take each character from str1
    found = False       # Assume not found in str2
    
    for j in str2:      # Compare with each character in str2
        if i == j:
            found = True
            break       # Stop checking once found
    
    if not found:       # If character not in str2
        rm += i

print(rm)
````


### 🚀 Cleaner Two Loop Version (Same Logic)


````python
str1 = "roshan"
str2 = "har"

rm = ""

for i in str1:
    for j in str2:
        if i == j:
            break
    else:        # Runs only if loop didn't break
        rm += i

print(rm)
````

# 🚩 Problem Statement: Leaders in an Array



Given an array of integers, print all the **leaders** in the array.

* An element is considered a **leader** if it is **greater than or equal to** all the elements to its right.
* The **rightmost element** is always a leader.

**Input Format**
* The first line contains an integer $N$, the size of the array.
* The second line contains $N$ space-separated integers.

**Output Format**
* Print all the leaders in the array in the order they appear (from left to right).

**Constraints**
* $1 \leq N \leq 10^5$
* $-10^9 \leq \text{arr}[i] \leq 10^9$

---
**📝 Sample Cases**

**Sample Input 1**
```text
6
16 17 4 3 5 2
```
**Sample Output 1:** `17 5 2`

**Explanation (Right to Left):**
* **2** is the rightmost element $\rightarrow$ **Leader**.
* **5** is $> 2$ $\rightarrow$ **Leader**.
* **3** is not $\geq 5$.
* **4** is not $\geq 5$.
* **17** is $> 5$ $\rightarrow$ **Leader**.
* **16** is not $\geq 17$.

**Sample Input 2**
```text
5
7 10 4 10 6
```
**Sample Output 2:** `10 10 6`

**Explanation (Right to Left):**
* **6** $\rightarrow$ **Leader**.
* **10** is $> 6$ $\rightarrow$ **Leader**.
* **4** is not $\geq 10$.
* **10** is $\geq 10$ $\rightarrow$ **Leader**.
* **7** is not $\geq 10$.


**🐍 Python Implementation**
This approach uses a **Right-to-Left** scan to maintain an $O(N)$ time complexity. We track the maximum value seen from the right; if a current element is greater than or equal to that maximum, it is a leader.

````python
def findLeaders(arr):
    n = len(arr)
    leaders = []
    
    # The rightmost element is always a leader
    max_from_right = arr[-1]
    leaders.append(max_from_right)
    
    # Iterate from right to left (second to last element down to index 0)
    for i in range(n - 2, -1, -1):
        if arr[i] >= max_from_right:
            max_from_right = arr[i]
            leaders.append(arr[i])
            
    # Reverse the list to maintain the original order of appearance
    return leaders[::-1]

# Driver Code
if __name__ == "__main__":
    # Reading input
    n = int(input().strip())
    arr = list(map(int, input().split()))
    
    # Finding and printing results
    result = findLeaders(arr)
    print(*result)
````

# 🚩 Problem: Missing Number in an Array



**Problem Statement**

You are given an array containing $n$ distinct numbers taken from the range **0 to n**. Since there are $n + 1$ numbers in the range and only $n$ numbers in the array, exactly one number is missing. Your task is to find and print that missing number.

**Input Format**

  * The first line contains an integer $n$, the size of the array.
  * The second line contains $n$ space-separated distinct integers in the range **0 to n**.

**Output Format**

  * Print the missing number.

**Constraints**

  * $1 \leq n \leq 10^5$
  * $0 \leq \text{arr}[i] \leq n$
  * All elements are distinct.



**🐍 Python Implementation**
The logic here is quite clever (and mathematically satisfying): we calculate what the sum *should* be for all numbers from 0 to $n$ and subtract the actual sum of the array. The difference is our missing number\!

````python
# Read the size of the array
n = int(input().strip())

# Read the space-separated integers into a list
arr = list(map(int, input().split()))

# Calculate the expected sum of numbers from 0 to n using the formula: n*(n+1)/2
expected_sum = n * (n + 1) // 2

# Calculate the actual sum of elements present in the array
actual_sum = sum(arr)

# The missing number is the difference
print(expected_sum - actual_sum)
````

# 🚩 Problem: Merge Overlapping Intervals


**Problem Statement**

You are given an array of intervals where: **intervals[i] = [start$_i$, end$_i$]**.
Your task is to merge all overlapping intervals and return a list of non-overlapping intervals that cover all the intervals in the input.

**Input Format**
* The first line contains an integer **N**, representing the number of intervals.
* The next **N** lines contain two space-separated integers **start end**, representing each interval.

**Output Format**
* Print the merged intervals, one per line.
* Each line should contain two space-separated integers representing the merged interval.

**Constraints**
* $1 \leq N \leq 10^5$
* $-10^9 \leq \text{start, end} \leq 10^9$
* $\text{start} \leq \text{end}$


**📝 Sample Case**

** Sample Input 1**
```text
4
1 3
2 6
8 10
15 18
```

**Sample Output 1**
```text
1 6
8 10
15 18
```

**Explanation:**
The intervals **[1, 3]** and **[2, 6]** overlap because the second interval starts before the first one ends ($2 \leq 3$). They are merged into **[1, 6]**. The other intervals do not overlap.



## 🐍 Python Implementation
This code uses a sorting-based approach. By sorting the intervals by their start times, we only need to check the current interval against the "running" merged interval to decide whether to merge them or start a new one.


````python
def mergeIntervals(intervals):
    if not intervals:
        return []
    
    # Step 1: Sort intervals by start time
    intervals.sort()
    
    merged = []
    start, end = intervals[0]
    
    for i in range(1, len(intervals)):
        curr_start, curr_end = intervals[i]
        
        # Overlapping case: current starts before/at the end of previous
        if curr_start <= end:
            end = max(end, curr_end)
        else:
            # No overlap, push the completed interval and start a new one
            merged.append([start, end])
            start, end = curr_start, curr_end
            
    # Don't forget to append the last interval
    merged.append([start, end])
    return merged

# Driver Code
if __name__ == "__main__":
    n = int(input().strip())
    intervals = []
    
    for i in range(n):
        s, e = map(int, input().split())
        intervals.append([s, e])
    
    result = mergeIntervals(intervals)
    
    for s, e in result:
        print(s, e)
````
