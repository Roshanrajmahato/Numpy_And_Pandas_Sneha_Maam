# Python Problem: Sort a List Using Bubble Sort

## Problem Statement
Write a Python program to **sort a list of numbers using Bubble Sort technique**.

Bubble Sort works by **comparing elements and swapping them** if they are in the wrong order.

**Input**  
l1 = [8, 4, 2, 1, 7]

**Output**

Before sorting => [8, 4, 2, 1, 7]
After sorting  -> [1, 2, 4, 7, 8]


---

## Explanation
- Bubble Sort repeatedly compares elements.
- If the current element is greater than the next element, they are swapped.
- This process continues until the list is completely sorted.
- After every pass, the smallest elements move toward the beginning.

Example:
Original list → [8, 4, 2, 1, 7]  
After sorting → [1, 2, 4, 7, 8]

---

## Algorithm Steps (HHL — Human High-Level Logic)

- Start  
- Take a list of numbers.  
- Begin comparing numbers from the start of the list.  
- Compare one number with the next number.  
- If the first number is greater than the next number, swap their positions.  
- Continue comparing all numbers in the list.  
- Repeat the comparison process multiple times.  
- With each repetition, larger numbers move toward their correct positions.  
- Continue until no more swaps are needed.  
- Display the sorted list.  
- Stop.  


## Python Code (Your Approach)

```python

def bubble_sort(l1):

    for i in range(len(l1)):

        for j in range(i + 1, len(l1)):

            if l1[i] > l1[j]:

                l1[i], l1[j] = l1[j], l1[i]


l1 = [8, 4, 2, 1, 7]

print("Before sorting =>", l1)

bubble_sort(l1)

print("After sorting ->", l1)
```

## Improved Version (Standard Bubble Sort — Recommended)

```python
def bubble_sort(l1):

    n = len(l1)

    for i in range(n):

        for j in range(0, n - i - 1):

            if l1[j] > l1[j + 1]:

                l1[j], l1[j + 1] = l1[j + 1], l1[j]


l1 = [8, 4, 2, 1, 7]

print("Before sorting =>", l1)

bubble_sort(l1)

print("After sorting ->", l1)
```

## Alternative Optimized Version (With Swap Check)

```python
def bubble_sort(l1):

    n = len(l1)

    for i in range(n):

        swapped = False

        for j in range(0, n - i - 1):

            if l1[j] > l1[j + 1]:

                l1[j], l1[j + 1] = l1[j + 1], l1[j]

                swapped = True

        if not swapped:
            break


l1 = [8, 4, 2, 1, 7]

print("Before sorting =>", l1)

bubble_sort(l1)

print("After sorting ->", l1)
```

# Python Problem: Sort a List Using Selection Sort

## Problem Statement
Write a Python program to **sort a list of numbers using Selection Sort technique**.

Selection Sort works by **selecting the smallest element** from the unsorted portion of the list and placing it at the correct position.

**Input**  
l1 = [8, 4, 2, 1, 7]

**Output**

Before sorting -> [8, 4, 2, 1, 7]
After sorting  -> [1, 2, 4, 7, 8]


---

## Explanation
- Selection Sort divides the list into **sorted** and **unsorted** parts.
- In each step:
  - Find the **minimum element** from the unsorted portion.
  - Swap it with the first unsorted element.
- Repeat this until the list becomes sorted.

Example:
Original → [8, 4, 2, 1, 7]  
After sorting → [1, 2, 4, 7, 8]

---

## Algorithm Steps (HHL — Human High-Level Logic)

- Start  
- Take a list of numbers.  
- Begin from the first position in the list.  
- Assume the current position contains the smallest element.  
- Compare this element with all remaining elements in the list.  
- If a smaller element is found, remember its position.  
- After checking all remaining elements, swap the smallest element with the current position.  
- Move to the next position in the list.  
- Repeat the process until all elements are sorted.  
- Display the sorted list.  
- Stop.  


## Python Code (Corrected Version of Your Code)

```python
def selection_sort(l1):

    for i in range(len(l1)):

        min = i

        for j in range(i + 1, len(l1)):

            if l1[j] < l1[min]:

                min = j

        l1[i], l1[min] = l1[min], l1[i]


l1 = [8, 4, 2, 1, 7]

print("Before sorting ->", l1)

selection_sort(l1)

print("After sorting ->", l1)
```

## Alternative Version (Using Different Variable Name)

```python
def selection_sort(l1):

    n = len(l1)

    for i in range(n):

        min_index = i

        for j in range(i + 1, n):

            if l1[j] < l1[min_index]:

                min_index = j

        l1[i], l1[min_index] = l1[min_index], l1[i]


l1 = [8, 4, 2, 1, 7]

print("Before sorting ->", l1)

selection_sort(l1)

print("After sorting ->", l1)
```
