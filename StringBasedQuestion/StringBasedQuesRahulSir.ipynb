# Python Problem: Check Whether a String is a Palindrome

## Problem Statement
Write a Python program to **check whether a given string is a palindrome**.

A **palindrome** is a word that reads the same **forward and backward**.

**Input**  
s1 = "MadaM"

**Output**  
Palindrome

---

## Explanation
- A palindrome string remains the same when reversed.
- The program compares characters from the **beginning** and **end** of the string.
- If all characters match, the string is a palindrome.
- If any character does not match, it is not a palindrome.
- In your example, `"MadaM"` is treated as palindrome because characters match exactly (case-sensitive).

Example:
"MadaM" → Reverse → "MadaM" → Same → Palindrome  

---

## Algorithm Steps (HHL — Human High-Level Logic)

- Start  
- Take a string as input.  
- Identify the first character and the last character of the string.  
- Compare the first and last characters.  
- If they are not equal, declare the string as **not a palindrome** and stop.  
- If they are equal, move one step forward from the beginning and one step backward from the end.  
- Repeat the comparison of characters from both ends.  
- Continue this process until all characters are checked.  
- If all characters match, declare the string as a **palindrome**.  
- Stop.  


## Python Code (Your Method — Two Pointer Approach)

````python
s1 = "MadaM"
j = len(s1) - 1

for i in range(len(s1)):
    if s1[i] != s1[j]:
        print("Not a palindrome")
        break
    j -= 1
else:
    print("Palindrome")
````

## Optimized Python Code (Using Reverse Slicing — Recommended)

````python
s1 = "MadaM"
if s1 == s1[::-1]:
    print("Palindrome")
else:
    print("Not palindrome")
````

## Alternative Approach (Case-Insensitive Palindrome)

````python
s1 = "MadaM"
s1 = s1.lower()
if s1 == s1[::-1]:
    print("Palindrome")
else:
    print("Not palindrome")
````

## Alternative Variation (Ignoring Spaces)

````python
s1 = "nurses run"
s1 = s1.replace(" ", "").lower()
if s1 == s1[::-1]:
    print("Palindrome")
else:
    print("Not palindrome")
````

# Python Problem: Combine Two Strings by Alternating Characters (Second String Reversed)

## Problem Statement
Write a Python program to **combine two strings** such that:
- Take characters from the **first string from left to right**
- Take characters from the **second string from right to left**
- Combine them alternately to form a new string

**Input**  
s1 = "abcd"  
s2 = "wxyz"

**Output**  
`azbycxdw`

---

## Explanation
- Characters from the first string are taken **in normal order**.
- Characters from the second string are taken **in reverse order**.
- Each character from `s1` is combined with the corresponding reversed character from `s2`.
- The result is stored in a new string.

Step-by-step:
- s1[0] → a  
- s2[last] → z → "az"  
- s1[1] → b  
- s2[last-1] → y → "azby"  
- Continue until all characters are used.

Final Output:
`azbycxdw`

---

## Algorithm Steps (HHL — Human High-Level Logic)

- Start  
- Take two strings of equal length.  
- Create an empty string to store the result.  
- Begin from the first character of the first string.  
- At the same time, begin from the last character of the second string.  
- Take one character from the first string.  
- Take one character from the second string starting from the end.  
- Join both characters and add them to the result string.  
- Move forward in the first string and backward in the second string.  
- Repeat this process until all characters are combined.  
- Display the final combined string.  
- Stop.  


````python
s1 = "abcd"
s2 = "wxyz"
s3 = ""
for i in range(len(s1)):
    s3 += s1[i] + s2[len(s2) - 1 - i]
print(s3)
````

## Optimized Python Code (Using zip — Recommended)

````python
s1 = "abcd"
s2 = "wxyz"
s3 = ""
for a, b in zip(s1, reversed(s2)):
    s3 += a + b
print(s3)
````

## Alternative Approach (Using List Join — More Efficient)

````python

s1 = "abcd"
s2 = "wxyz"
result = []
for i in range(len(s1)):
    result.append(s1[i])
    result.append(s2[len(s2)-1-i])
s3 = "".join(result)
print(s3)
````

## Alternative Variation (Handle Different Length Strings)

````python
s1 = "abc"
s2 = "wxyz"
s3 = ""
min_len = min(len(s1), len(s2))
for i in range(min_len):
    s3 += s1[i] + s2[len(s2)-1-i]
s3 += s1[min_len:] + s2[:len(s2)-min_len]

print(s3)
````

# Python Problem: Check Whether Two Strings are Anagrams

**Problem Statement**
Write a Python program to **check whether two given strings are anagrams**.

Two strings are called **anagrams** if:
- They contain the **same characters**
- With the **same frequency**
- Possibly in a **different order**

**Input**  
s1 = "Silent"  
s2 = "listen"

**Output**  
Anagram string  

---

**Explanation**
- The program sorts the characters of both strings.
- After sorting:
  - If both sorted strings are equal → They are anagrams.
  - Otherwise → They are not anagrams.
- In this example:
  - "Silent" → sorted → "Seilnt"  
  - "listen" → sorted → "eilnst"  
- Since uppercase **S** and lowercase **s** are different, this comparison is **case-sensitive**.
- If converted to lowercase first, both become equal → Anagram.

---

**Algorithm Steps (HHL — Human High-Level Logic)**

- Start  
- Take two strings as input.  
- Convert both strings into lists of characters.  
- Arrange the characters of each string in alphabetical order.  
- After sorting, form the strings again from the sorted characters.  
- Compare the sorted versions of both strings.  
- If both sorted strings are the same, declare them as **anagram strings**.  
- Otherwise, declare them as **not anagram strings**.  
- Stop.  


## Python Code (Your Approach — Manual Sorting)

````python
def sort_the_string(s):
    l = list(s)

    for i in range(len(l)):
        for j in range(i + 1, len(l)):

            if l[i] > l[j]:
                l[i], l[j] = l[j], l[i]

    return "".join(l)

s1 = "Silent"
s2 = "listen"

if sort_the_string(s1) == sort_the_string(s2):
    print("Anagram string")
else:
    print("Not a anagram")
````

## Improved Version (Case-Insensitive — Recommended)

````python
def sort_the_string(s):

    l = list(s.lower())

    for i in range(len(l)):
        for j in range(i + 1, len(l)):

            if l[i] > l[j]:
                l[i], l[j] = l[j], l[i]

    return "".join(l)


s1 = "Silent"
s2 = "listen"

if sort_the_string(s1) == sort_the_string(s2):
    print("Anagram string")
else:
    print("Not an anagram")
````

## Best Optimized Approach (Using Built-in sorted())

````python
s1 = "Silent"
s2 = "listen"
if sorted(s1.lower()) == sorted(s2.lower()):
    print("Anagram string")
else:
    print("Not an anagram")
````

## Alternative Approach (Using Character Count — Most Efficient)

````python
s1 = "Silent"
s2 = "listen"
if len(s1) != len(s2):
    print("Not an anagram")
else:
    d1 = {}
    d2 = {}
    for ch in s1.lower():
        d1[ch] = d1.get(ch, 0) + 1
    for ch in s2.lower():
        d2[ch] = d2.get(ch, 0) + 1
    if d1 == d2:
        print("Anagram string")
    else:
        print("Not an anagram")
````

## Alternative Variation (Ignoring Spaces)

````python
s1 = "listen"
s2 = "silent"
s1 = s1.replace(" ", "").lower()
s2 = s2.replace(" ", "").lower()
if sorted(s1) == sorted(s2):
    print("Anagram string")
else:
    print("Not an anagram")
````

# Python Problem: Check Whether a String is a Pangram



**Problem Statement**
Write a Python program to **check whether a given string is a pangram**.

A **pangram** is a sentence that contains **all 26 letters of the English alphabet at least once**.

**Input**  
s1 = "Pack my box with five dozen liquor jugs"

**Output**  
Pangram string  

---

**Explanation**
- A pangram must contain **every alphabet letter from 'a' to 'z'**.
- The program reads each character of the string.
- Spaces are ignored because they are not letters.
- All letters are converted to lowercase to ensure case-insensitive comparison.
- Unique letters are stored in a set.
- If the set contains **26 unique letters**, the string is a pangram.

Example:
"Pack my box with five dozen liquor jugs"  
Contains all letters → Pangram  

---

**Algorithm Steps (HHL — Human High-Level Logic)**

- Start  
- Take a sentence as input.  
- Create an empty collection to store unique letters.  
- Read each character from the sentence one by one.  
- If the character is a space, ignore it and move to the next character.  
- Convert the character into lowercase form.  
- Store the character if it is not already present in the collection.  
- Continue this process until all characters are checked.  
- Count the number of unique letters stored.  
- If the count equals 26, declare the sentence as a **pangram string**.  
- Otherwise, declare it as **not a pangram string**.  
- Stop.  


## Python Code (Your Approach)

````python
# s1 = "The quick brown fox jumps over a lazy dog"
s1 = "Pack my box with five dozen liquor jugs"

s2 = set()

for i in s1:

    if i == " ":
        continue

    elif i not in s2:
        s2.add(i.lower())

if len(s2) == 26:
    print("Panagram string")
else:
    print("Not a panagram string")
````

## Improved Version (Cleaner Logic — Recommended)

````python
s1 = "Pack my box with five dozen liquor jugs"

letters = set()

for ch in s1.lower():

    if ch.isalpha():
        letters.add(ch)

if len(letters) == 26:
    print("Pangram string")
else:
    print("Not a pangram string")
````

## Best Optimized Approach (Using Alphabet Set)

````python
import string

s1 = "Pack my box with five dozen liquor jugs"
alphabet_set = set(string.ascii_lowercase)
if alphabet_set.issubset(set(s1.lower())):
    print("Pangram string")
else:
    print("Not a pangram string")
````

## Alternative Variation (Display Missing Letters)

````python
import string

s1 = "Pack my box with five dozen liquor jugs"

letters = set(s1.lower())

missing = set(string.ascii_lowercase) - letters

if len(missing) == 0:
    print("Pangram string")
else:
    print("Missing letters:", missing)
````

# Python Problem: Check Whether a String is a Mirror String


**Problem Statement**

Write a Python program to **check whether a given string is a mirror string**.

A **mirror string** is a string where **every character looks the same when viewed in a mirror**.

Valid mirror characters include:
A, H, I, M, O, T, U, V, W, X, Y  
(and some lowercase forms like o, v, w, x, 8)

**Input**  
s1 = "MOM"

**Output**  
Mirror String  

---

**Explanation**
- Some characters remain unchanged when seen in a mirror.
- The program checks each character in the string.
- If any character is **not a mirror character**, the string is not a mirror string.
- If all characters belong to the mirror character set, it is declared a mirror string.

Example:
"MOM" → M and O are mirror characters → Mirror String  

---

**Algorithm Steps (HHL — Human High-Level Logic)**

- Start  
- Take a string as input.  
- Prepare a collection of characters that remain unchanged in a mirror.  
- Read each character of the string one by one.  
- Check whether the character belongs to the mirror character collection.  
- If any character does not belong to the collection:  
   - Declare the string as **not a mirror string**.  
   - Stop the process.  
- If all characters belong to the mirror character collection:  
   - Declare the string as a **mirror string**.  
- Stop.  


## Python Code (Your Approach)

````python
s1 = 'AHA'

def mirror_str(s1):

    mirror_chars = {
        'A','H','I','M','O',
        'T','U','V','W','X',
        'Y','o','v','w','x','8'
    }

    for i in s1:

        if i not in mirror_chars:
            return False

    return True


s1 = "MOM"

if mirror_str(s1):
    print("Mirror String")
else:
    print("Not a mirror string")
````

## Improved Version (Case-Insensitive — Recommended)

````python
def mirror_str(s):

    mirror_chars = {
        'A','H','I','M','O',
        'T','U','V','W','X','Y'
    }

    s = s.upper()

    for ch in s:

        if ch not in mirror_chars:
            return False

    return True


s1 = "Mom"

if mirror_str(s1):
    print("Mirror String")
else:
    print("Not a mirror string")
````

## Alternative Approach (Using all())

````python
def mirror_str(s):

    mirror_chars = {
        'A','H','I','M','O',
        'T','U','V','W','X','Y'
    }

    return all(ch.upper() in mirror_chars for ch in s)


s1 = "MOM"

if mirror_str(s1):
    print("Mirror String")
else:
    print("Not a mirror string")
````

## Alternative Variation (Check Perfect Mirror Symmetry)

````python
mirror_map = {
    'A':'A','H':'H','I':'I',
    'M':'M','O':'O','T':'T',
    'U':'U','V':'V','W':'W',
    'X':'X','Y':'Y'
}

s1 = "MOM"

flag = True

for i in range(len(s1)):

    if s1[i] != mirror_map.get(s1[len(s1)-1-i], ''):
        flag = False
        break

if flag:
    print("Mirror Symmetric String")
else:
    print("Not Mirror Symmetric")
````

# Python Problem: Print String in Square Root Pattern (Encryption Pattern)



**Problem Statement**
Write a Python program to **arrange the characters of a string into multiple lines** based on the **square root of the string length** (after removing spaces).

This type of pattern is commonly used in **simple text encryption formatting**.

**Input**  
Enter the string:  
have a nice day  

**Output**
```
have
anic
eday
```

---

**Explanation**
- First, remove all spaces from the string.
- Count the total number of characters remaining.
- Find the square root of the length.
- Round the result **upwards** to ensure all characters fit.
- Print characters in groups based on this calculated size.
- Each group forms one line.

Example:
String = "have a nice day"  
After removing spaces → "haveaniceday"  
Length = 12  
Square root ≈ 3.46 → Rounded up = 4  

Characters printed in groups of 4:
have  
anic  
eday  

---

**Algorithm Steps (HHL — Human High-Level Logic)**

- Start  
- Take a string as input from the user.  
- Remove all spaces from the string.  
- Count the number of characters remaining.  
- Find the square root of the character count.  
- Round the square root value upward to get the number of characters per line.  
- Begin reading characters from the string.  
- Print characters in groups equal to the calculated line size.  
- Move to the next group of characters.  
- Continue until all characters are printed.  
- Stop.  


````python
import math

# Input string
s = input("Enter the string: ")
# Step 1: Remove spaces
s_no_space = s.replace(" ", "")
# Step 2: Count characters
length = len(s_no_space)
# Step 3: Find square root (ceil to avoid missing chars)
line_size = math.ceil(math.sqrt(length))
print("Length (without spaces):", length)
print("Characters per line:", line_size)
print("\nOutput:")
# Step 4: Print in chunks
for i in range(0, length, line_size):
    print(s_no_space[i:i+line_size])
````

## Alternative Variation (Return List Instead of Printing)

````python
import math
s = input("Enter the string: ")
s = s.replace(" ", "")
length = len(s)
line_size = math.ceil(math.sqrt(length))
result = []
for i in range(0, length, line_size):
    result.append(s[i:i+line_size])
print(result)
````

# Python Problem: Check Whether a Password is Valid



## Problem Statement
Write a Python program to **check whether a given password is valid** based on the following rules:

### Password Rules
- Password length must be **at least 8 characters** and **at most 30 characters**  
- Must contain **at least one uppercase letter**  
- Must contain **at least one lowercase letter**  
- Must contain **at least one digit**  
- Must contain **at least one special character**

**Input**  
Enter the password:  
Abc@1234  

**Output**  
It can be a valid password  

---

## Explanation
- The program first checks the **length** of the password.
- Then it checks each character one by one.
- It counts:
  - Uppercase letters  
  - Lowercase letters  
  - Numbers  
  - Special characters  
- If all required types are present, the password is valid.

Example:
Password = **Abc@1234**  
Contains:
- Uppercase → A  
- Lowercase → bc  
- Digit → 1234  
- Special Character → @  

So it is a **valid password**.

---

## Algorithm Steps (HHL — Human High-Level Logic)

- Start  
- Take a password as input from the user.  
- Count the total number of characters in the password.  
- Check whether the length is between **8 and 30 characters**.  
- If the length is not within this range, declare it invalid.  
- Otherwise, begin checking each character in the password.  
- Count lowercase letters present in the password.  
- Count uppercase letters present in the password.  
- Count numeric digits present in the password.  
- Count special characters present in the password.  
- After checking all characters, verify whether:  
    - At least one lowercase letter exists.  
    - At least one uppercase letter exists.  
    - At least one digit exists.  
    - At least one special character exists.  
- If all conditions are satisfied, declare the password as valid.  
- Otherwise, declare the password as invalid.  
- Stop.  


## Python Code (Corrected Version of Your Code)

````python
s1 = input("Enter the password: ")
# Check length condition
if len(s1) >= 8 and len(s1) <= 30:
    uc_c, lc_c, n_c, s_c = 0, 0, 0, 0
    for i in s1:
        if i >= 'a' and i <= 'z':
            lc_c += 1
        elif i >= 'A' and i <= 'Z':
            uc_c += 1
        elif i >= '0' and i <= '9':
            n_c += 1
        else:
            s_c += 1
    if lc_c >= 1 and uc_c >= 1 and n_c >= 1 and s_c >= 1:
        print("It can be a valid password")
    else:

        print("It cannot be a valid password")

else:

    print("It should contain at least 8 and at most 30 characters")
````

## Improved Version (Using Built-in Functions — Recommended)

````python
s1 = input("Enter the password: ")
if 8 <= len(s1) <= 30:
    has_upper = any(ch.isupper() for ch in s1)
    has_lower = any(ch.islower() for ch in s1)
    has_digit = any(ch.isdigit() for ch in s1)
    has_special = any(not ch.isalnum() for ch in s1)
    if has_upper and has_lower and has_digit and has_special:
        print("It can be a valid password")
    else:
        print("It cannot be a valid password")
else:
    print("Password length must be between 8 and 30 characters")
````

Write a Python program to **arrange characters of a string into rows** based on the **square root of the number of non-space characters**, while still **printing spaces in the output**.

**Input**  
s1 = "asdfjadjfbawdbfwfiwifiqw qwiogfniqwgfouqwiugbfhqwf qwrfbqwubfuqwufiuqwlfhjwf"

**Output (Pattern Example)**
```

asdfjadjfb
awdbfwfiwi

fniqwgfouq
wiugbfhqwf
qwrfbqwub
fuqwufiuqw
lfhjwf

```

---

**Explanation**
- First, count only **non-space characters**.
- Find the **square root** of the count.
- Round the value to determine how many characters should appear in each row.
- Then print characters line by line.
- Spaces are included while printing but **not counted** in square root calculation.

This creates a **grid-style string layout**.

---

**Algorithm Steps (HHL — Human High-Level Logic)**

- Start  
- Take a string as input.  
- Initialize a counter to count characters excluding spaces.  
- Read each character in the string.  
- If the character is a space, ignore it.  
- Otherwise, increase the counter value.  
- After counting characters, find the square root of the count.  
- Round the square root value to determine characters per row.  
- Begin reading characters again from the string.  
- Print characters one by one.  
- After printing the required number of characters, move to the next line.  
- Continue until all characters are printed.

### Python Code (Your Approach)

````python

import math

s1 = "asdfjadjfbawdbfwfiwifiqw qwiogfniqwgfouqwiugbfhqwf qwrfbqwubfuqwufiuqwlfhjwf"

# Count characters excluding spaces
cnt = 0

for i in s1:

    if i == " ":
        continue

    else:
        cnt += 1
# Find square root value
val = round(math.sqrt(cnt))

# Create formatted output
s2 = ""
c = 0
for i in s1:

    if c == val:
        s2 += "\n"
        c = 0

    s2 += i

    c += 1


print(s2)
````

````python
import math

s1 = input("Enter the string: ")

# Count non-space characters
cnt = 0

for ch in s1:

    if ch != " ":
        cnt += 1


# Use ceil instead of round
val = math.ceil(math.sqrt(cnt))


c = 0

for ch in s1:

    if c == val:
        print()
        c = 0

    print(ch, end="")

    c += 1
````

# Python Problem: Extract Alphanumeric Characters and Reverse Them


**Problem Statement**

Write a Python program to **extract only alphanumeric characters**  
(**letters and digits**) from a given list or string and then **reverse the extracted elements**.

Alphanumeric characters include:
- Uppercase letters (A–Z)  
- Lowercase letters (a–z)  
- Digits (0–9)

**Input**  
l1 = ['a', '#', 'B', '7', '@', 'c', '9', '!']

**Output**
```

Extracted List  -> ['a', 'B', '7', 'c', '9']
Reversed List   -> ['9', 'c', '7', 'B', 'a']

```

---

**Explanation**
- The program scans each character.
- It checks whether the character is:
  - a lowercase letter  
  - an uppercase letter  
  - a digit  
- If yes, it stores the character.
- After collecting all valid characters, the list is reversed.

This removes **special characters** and keeps only **letters and numbers**.

---

**Algorithm Steps (HHL — Human High-Level Logic)**

- Start  
- Take a list or string as input.  
- Initialize a counter to count valid characters.  
- Read each element one by one.  
- Check whether the element is:
   - a lowercase letter, or  
   - an uppercase letter, or  
   - a digit.  - If the element satisfies any condition, increase the counter.  
- Create a new list of size equal to the counted valid elements.  
- Read the original list again.  
- Store only valid characters into the new list.  
- Reverse the new list.  
- Display the reversed list.  
- Stop.  


### Python Code (Corrected Version — Your Logic Style)

````python
l1 = ['a', '#', 'B', '7', '@', 'c', '9', '!']

cnt = 0

# Count alphanumeric characters
for i in range(len(l1)):

    if (ord(l1[i]) >= 97 and ord(l1[i]) <= 122) or \
       (ord(l1[i]) >= 65 and ord(l1[i]) <= 90) or \
       (ord(l1[i]) >= 48 and ord(l1[i]) <= 57):

        cnt += 1


# Create new list
l2 = [0] * cnt

print("Initial list:", l2)

j = 0

# Store valid characters
for i in range(len(l1)):

    if (ord(l1[i]) >= 97 and ord(l1[i]) <= 122) or \
       (ord(l1[i]) >= 65 and ord(l1[i]) <= 90) or \
       (ord(l1[i]) >= 48 and ord(l1[i]) <= 57):

        l2[j] = l1[i]
        j += 1


print("Extracted List ->", l2)

# Reverse list
l2 = l2[::-1]

print("Reversed List ->", l2)
````

## Improved Version (Using Built-in Function — Recommended)

````python
l1 = ['a', '#', 'B', '7', '@', 'c', '9', '!']

l2 = []

for ch in l1:

    if ch.isalnum():

        l2.append(ch)


print("Extracted List ->", l2)

l2.reverse()

print("Reversed List ->", l2)
````

# Python Problem: Reverse Only Alphanumeric Characters in a List

Write a Python program to **reverse only alphanumeric characters**  
(**letters and digits**) in a list while keeping **special characters in their original positions**.

Alphanumeric characters include:
- Uppercase letters (A–Z)  
- Lowercase letters (a–z)  
- Digits (0–9)

**Input**  
l1 = ['L', '1', '$', 'u', '#', '7', '8', 'h', '3', 'a', '2', '@', 'r', '5', 'r']

**Output**

Original List  -> ['L', '1', '$', 'u', '#', '7', '8', 'h', '3', 'a', '2', '@', 'r', '5', 'r']

Modified List  -> ['r', '5', '$', 'r', '#', '2', 'a', '3', 'h', '8', '7', '@', 'u', '1', 'L']

  - One from the beginning.
  - One from the end.

---

## Algorithm Steps (HHL — Human High-Level Logic)

- Take a list containing characters.  
- Create a function to check whether a character is alphanumeric.  
- Set one pointer at the beginning of the list.

- Swap the two alphanumeric characters.  
- Move both pointers toward each other.

- Display the modified list.

# Python Problem: Fill a Matrix with Characters of a String (Repeating Pattern)



## Problem Statement
Write a Python program to **fill an N × N matrix using characters of a string**.  
If the string ends before filling the matrix, **repeat the string from the beginning**.

**Input**  
s1 = "PySpIdErS"  
Enter a value for matrix:  
3  

**Output**
```

P y S
p I d
E r S

```

If matrix size is larger than string length, the string repeats.

Example (n = 4):
```

P y S p
I d E r
S P y S
p I d E

```

---

## Explanation
- The string characters are printed into a matrix row by row.
- A counter (index) keeps track of which character to print.
- If the index reaches the string length, it resets to **0**.
- This allows the string to repeat until the matrix is fully filled.

Example:
String = "PySpIdErS"  
Matrix size = 3  

Total cells = 9  
String length = 9  
So characters fill exactly once.

---

## Algorithm Steps (HHL — Human High-Level Logic)

- Start  
- Take a string containing characters.  
- Take an integer value representing matrix size.  
- Initialize a variable to track the position of characters in the string.  
- Begin filling rows of the matrix one by one.  
- For each row, fill columns from left to right.  
- Print the current character from the string.  
- Move to the next character position.  
- If the end of the string is reached, restart from the first character.  
- Continue this process until all rows and columns are filled.  
- Move to the next line after completing each row.  
- Stop.  


## Python Code (Your Approach)

````python
s1 = "PySpIdErS"

n = int(input("Enter a value for matrix: "))

idx = 0

for i in range(n):

    for j in range(n):

        print(s1[idx], end=" ")

        idx += 1

        if idx >= len(s1):
            idx = 0

    print()
````

# Python Problem: Print Cross Pattern Using Characters of a String


## Problem Statement
Write a Python program to **print a cross ( + ) pattern** using characters of a string.  
This pattern works **only when the string length is an odd number**.

**Input**  
Enter a string:  
HELLO  

**Output**

```
    H
    E
H E L L O
    L
    O
```


---

## Explanation
- The program prints characters in the **middle row** and **middle column**.
- The middle index is calculated using **length // 2**.
- Characters are printed:
  - Across the **middle row**
  - Down the **middle column**
- All other positions remain blank.
- The pattern works only when the string length is **odd**, so there is a clear center.

Example:
String = HELLO  
Length = 5  
Middle Index = 2  

Row 2 → prints entire string  
Column 2 → prints characters vertically  

---

## Algorithm Steps (HHL — Human High-Level Logic)

- Start  
- Take a string as input from the user.  
- Find the total number of characters in the string.  
- Check whether the number of characters is odd.  
- If the number is even, display a message that the pattern works only for odd lengths.  
- If the number is odd, find the middle position of the string.  
- Begin creating rows equal to the string length.  
- For each row, create columns equal to the string length.  
- If the current row is the middle row, print all characters of the string.  
- If the current column is the middle column, print the corresponding character vertically.  
- Otherwise, print empty spaces.  
- Move to the next line after completing each row.  
- Continue until all rows are printed.  
- Stop.  


## Python Code (Corrected Version of Your Code)

````python
text = input("Enter a string: ")
n = len(text)
if n % 2 != 0:
    for i in range(n):
        for j in range(n):
            if i == n // 2:
                print(text[j], end=" ")
            elif j == n // 2:
                print(text[i], end=" ")
            else:
                print(" ", end=" ")
        print()
else:
    print("This pattern only works for odd number length string")
````


## Alternative Variation (Print X Pattern Instead of +)


````python
text = input("Enter a string: ")
n = len(text)
if n % 2 != 0:
    for i in range(n):
        for j in range(n):
            if i == j or j == n - i - 1:
                print(text[i], end=" ")
            else:
                print(" ", end=" ")
        print()
else:
    print("String length must be odd")
````
