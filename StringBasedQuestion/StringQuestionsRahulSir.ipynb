```python
s1="aaaaabbbbbccccaaaabbbbccdddddccccccc"
s2=""
for i in s1:
    count=0
    for j in s1:
        if i==j:
            count+=1
    if f"{i}{count}" not in s2:
        s2+=f"{i}{count}"
print(f"Compress String is {s2}") 
```
