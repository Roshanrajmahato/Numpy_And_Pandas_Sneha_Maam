# Python Problem: Check Palindrome using Recursion



**Problem Statement**

Write a Python program to **check whether a given string is a palindrome** using **recursion**.

A **palindrome** is a string that reads the **same forward and backward**.

The program should:

* Take a string input
* Use recursion to compare characters
* Return **True** if palindrome
* Return **False** otherwise

---

**Input**

```text
madam
```

---

**Output**

```text
True
```

---

## Explanation

A **palindrome** string remains the same when reversed.

Examples:

```text
madam  → Palindrome
racecar → Palindrome
hello  → Not Palindrome
```

### Recursive Logic

Compare:

* First character (`l`)
* Last character (`r`)

If equal:

* Move inward (`l+1`, `r-1`)

If indices cross:

* String is palindrome ✅

---

## Algorithm Steps

* Start
* Define recursive function `isPalindrome(s, l, r)`
* If `l >= r`, return True
* Check if characters at `l` and `r` are equal
* If equal, call recursion with `l+1`, `r-1`
* Otherwise return False
* Input string
* Call palindrome function
* Print result
* Stop

---

## Python Code

```python
def isPalindrome(s, l, r):

    # Base case
    if l >= r:
        return True

    # Recursive case
    return (s[l] == s[r] and 
            isPalindrome(s, l + 1, r - 1))


# Main execution
s = input()

print(isPalindrome(s, 0, len(s) - 1))
```


# Python Problem: Remove All Occurrences of `'a'` from String (Recursion)

## Problem Statement

Write a Python program to **remove all occurrences of character `'a'`** from a given string using **recursion**.

The program demonstrates **two recursive approaches**:

1. Using **index parameter**
2. Using **substring method**

---

## Input

```text
abacada
```

---

## Output

```text
bcd
bcd
```

---

## Explanation

We remove every `'a'` from the string recursively.

Example:

```text
Input  : abacada
Remove : a
Output : bcd
```

---

## Approach 1: Using Index

* Start from index `0`
* Process each character
* If character ≠ `'a'`, keep it
* Otherwise skip it
* Continue recursion

---

## Approach 2: Using Substring

* Take first character
* Recursively process remaining string
* If character ≠ `'a'`, add it
* Otherwise skip it

---

## Algorithm Steps

### Method 1 (Index Based)

* Start
* Define function `removeA(s, idx)`
* If `idx == len(s)` → return empty string
* Recursively call `removeA(s, idx+1)`
* If current character ≠ `'a'` → add to result
* Otherwise skip
* Return result

---

### Method 2 (Substring Based)

* Start
* Define function `removeA2(s)`
* If string length is `0` → return empty string
* Recursively call `removeA2(s[1:])`
* If first character ≠ `'a'` → add to result
* Otherwise skip
* Return result

---

## Python Code

```python
def removeA(s, idx):
    # Base case
    if idx == len(s):
        return ""

    # Recursive call
    smallAns = removeA(s, idx + 1)

    currChar = s[idx]

    if currChar != 'a':
        return currChar + smallAns
    else:
        return smallAns


def removeA2(s):
    # Base case
    if len(s) == 0:
        return ""

    # Recursive call
    smallAns = removeA2(s[1:])

    currChar = s[0]

    if currChar != 'a':
        return currChar + smallAns
    else:
        return smallAns


# Main execution
s = input()

print(removeA(s, 0))
print(removeA2(s))
```


# Python Problem: Reverse String using Recursion & Check Palindrome

## Problem Statement

Write a Python program to **reverse a string using recursion** and check whether the string is a **palindrome**.

A **palindrome** is a string that reads the **same forward and backward**.

The program should:

* Reverse the string using recursion
* Compare reversed string with original
* Print whether it is **Palindrome** or **Not Palindrome**

---

## Input

```text
madam
```

---

## Output

```text
madam is Palindrome
```

---

## Explanation

### Reverse Using Recursion

The recursive function:

* Moves forward through the string
* Builds reversed string while returning

Example:

```text
Input  : madam
Reverse Steps:
m → adam
a → dam
d → am
a → m
m → ""

Return Phase:
"" + m
m + a
ma + d
mad + a
mada + m

Output: madam
```

---

### Palindrome Check

After reversing:

```text
Original String == Reversed String
→ Palindrome
```

Otherwise:

```text
Not Palindrome
```

---

## Algorithm Steps

* Start
* Input string `s`
* Define recursive function `reverse(s, idx)`
* If `idx == len(s)` → return empty string
* Recursively call `reverse(s, idx+1)`
* Append current character
* Store reversed string
* Compare reversed string with original
* Print palindrome result
* Stop

---

## Python Code

```python
def reverse(s, idx):

    # Base case
    if idx == len(s):
        return ""

    # Recursive call
    smallAns = reverse(s, idx + 1)

    # Add current character at end
    return smallAns + s[idx]


# Main execution
s = input()

rev = reverse(s, 0)

if rev == s:
    print(f"{s} is Palindrome")
else:
    print(f"{s} is Not palindrome")
```


# Python Problem: Subsequence & Subset Sum using Recursion

## Problem Statement

Write Python programs to perform **subsequence generation** and **subset sum calculation** using **recursion**.

The program demonstrates:

1. Print all **subsequences** of a string
2. Print all **subset sums** of an array
3. Return all subsequences in a **list**

---

## Input

Array:

```text
[2, 4, 5]
```

String (example):

```text
"abc"
```

---

## Output

### Subset Sum Output

```text
0
5
4
9
2
7
6
11
```

### Subsequence Output (for "abc")

```text
abc
ab
ac
a
bc
b
c

```

### Subsequence List Output

```text
['', 'c', 'b', 'bc', 'a', 'ac', 'ab', 'abc']
```

---

# Explanation

## 1️⃣ Print Subsequences of String

At each step:

* Either **include** character
* Or **exclude** character

So total subsequences:

```text
2^n
```

---

## 2️⃣ Subset Sum

For each element:

* Either **add to sum**
* Or **skip**

This generates **all possible subset sums**.

---

## 3️⃣ Get Subsequences in List

Instead of printing:

* Store results in list
* Combine smaller answers

---

# Algorithm Steps

## printSSQ(s, currAns)

* If string empty → print current answer
* Take first character
* Call recursion including character
* Call recursion excluding character

---

## sumSubset(a, n, idx, sum)

* If index ≥ n → print sum
* Call recursion without including element
* Call recursion including element

---

## getSSQ(s)

* If string empty → return list with ""
* Get subsequences of remaining string
* Add:

  * Without character
  * With character

---

# Python Code

```python
# Print all subsequences
def printSSQ(s, currAns):

    # Base case
    if len(s) == 0:
        print(currAns)
        return

    currChar = s[0]
    remString = s[1:]

    # Include character
    printSSQ(remString, currAns + currChar)

    # Exclude character
    printSSQ(remString, currAns)


# Print subset sums
def sumSubset(a, n, idx, curr_sum):

    # Base case
    if idx >= n:
        print(curr_sum)
        return

    # Do not include element
    sumSubset(a, n, idx + 1, curr_sum)

    # Include element
    sumSubset(a, n, idx + 1, curr_sum + a[idx])


# Return subsequences as list
def getSSQ(s):

    ans = []

    # Base case
    if len(s) == 0:
        ans.append("")
        return ans

    curr = s[0]

    smallAns = getSSQ(s[1:])

    for ss in smallAns:
        ans.append(ss)
        ans.append(curr + ss)

    return ans


# Main execution

# Example array
a = [2, 4, 5]

print("Subset sums:")
sumSubset(a, len(a), 0, 0)


# Example string
s = "abc"

print("\nSubsequences:")
printSSQ(s, "")

print("\nSubsequence List:")
print(getSSQ(s))
```


# Python Problem: Print All Subsequences of a String (Recursion)

## Problem Statement

Write a Python program to **print all subsequences** of a given string using **recursion**.

A **subsequence** is formed by:

* Including a character
* Or excluding it

Total number of subsequences:

```text
2^n  (where n = length of string)
```

---

## Input

```text
abc
```

---

## Output

```text
abc
ab
ac
a
bc
b
c

```

(Note: Order may vary depending on recursion)

---

## Explanation

For string:

```text
"abc"
```

Each character has **two choices**:

* Include
* Exclude

So recursion forms:

```text
abc
ab
ac
a
bc
b
c
""
```

(`""` empty string is also a valid subsequence)

---

## Algorithm Steps

* Start
* Define function `printSSQ(s, currAns)`
* If string is empty → print current answer
* Take first character
* Create remaining string
* Call recursion including character
* Call recursion excluding character
* Stop

---

## Python Code

```python
def printSSQ(s, currAns):

    # Base case
    if len(s) == 0:
        print(currAns)
        return

    curr = s[0]
    remString = s[1:]

    # Include current character
    printSSQ(remString, currAns + curr)

    # Exclude current character
    printSSQ(remString, currAns)


# Main execution
printSSQ("abc", "")
```


# Python Problem: Return All Subsequences of a String (Recursion)

## Problem Statement

Write a Python program to **return all subsequences** of a given string using **recursion**.

Instead of printing subsequences, this program **stores them in a list** and returns the list.

---

## Input

```text
abc
```

---

## Output

```text
''
a
b
ab
c
ac
bc
abc
```

(Note: Order may vary depending on recursion)

---

## Explanation

For string:

```text
"abc"
```

Each character has **two choices**:

* Include the character
* Exclude the character

So total subsequences:

```text
2^n   (where n = length of string)
```

For `"abc"`:

```text
Subsequences:
""
"a"
"b"
"ab"
"c"
"ac"
"bc"
"abc"
```

---

## Algorithm Steps

* Start
* Define function `getSSQ(s)`
* If string is empty:

  * Return list containing `""`
* Take first character `curr`
* Recursively get subsequences of remaining string
* For each subsequence in smaller result:

  * Add subsequence
  * Add current character + subsequence
* Return final list
* Stop

---

## Python Code

```python
def getSSQ(s):

    ans = []

    # Base case
    if len(s) == 0:
        ans.append("")
        return ans

    curr = s[0]

    # Recursive call
    smallAns = getSSQ(s[1:])

    # Build result
    for ss in smallAns:
        ans.append(ss)          # without curr
        ans.append(curr + ss)   # with curr

    return ans


# Main execution
ans = getSSQ("abc")

for ss in ans:
    print(ss)
```


# Python Problem: Subset Sum using Recursion

## Problem Statement

Write a Python program to **print all possible subset sums** of a given array using **recursion**.

For each element in the array:

* Either **include** it in the sum
* Or **exclude** it

This generates **all possible subset sums**.

---

## Input

```text
[2, 4, 5]
```

---

## Output

```text
11
6
7
2
9
4
5
0
```

*(Order may vary depending on recursion)*

---

## Explanation

Given array:

```text
[2, 4, 5]
```

Possible subsets:

```text
{}
{2}
{4}
{5}
{2,4}
{2,5}
{4,5}
{2,4,5}
```

Corresponding sums:

```text
0
2
4
5
6
7
9
11
```

Each element has **two choices**:

```text
Include → Add element to sum  
Exclude → Skip element
```

Total subsets:

```text
2^n
```

---

## Algorithm Steps

* Start
* Define function `subsetSum(a, n, idx, sum)`
* If `idx >= n`:

  * Print `sum`
  * Return
* Call recursion including current element
* Call recursion excluding current element
* Stop

---

## Python Code

```python
def subsetSum(a, n, idx, curr_sum):

    # Base case
    if idx >= n:
        print(curr_sum)
        return

    # Include element
    subsetSum(a, n, idx + 1, curr_sum + a[idx])

    # Exclude element
    subsetSum(a, n, idx + 1, curr_sum)


# Main execution
a = [2, 4, 5]

subsetSum(a, len(a), 0, 0)
```

## Problem Statement

Write a Python program to solve the **Frog Jump problem** using **recursion**.

A frog is standing on the **first stone** and wants to reach the **last stone**.

Rules:

* Frog can jump **1 step** or **2 steps** forward
* Energy cost = **absolute difference** between heights
* Find the **minimum total energy** required to reach the last stone

---

## Input

```text
[10, 30, 40, 20]
```

---

## Output

```text
30
```

---

## Explanation

Given heights:

```text
[10, 30, 40, 20]
```

Possible paths:

### Path 1

```text
10 → 30 → 40 → 20

Cost:
|10−30| = 20  
|30−40| = 10  
|40−20| = 20  

Total = 50
```

### Path 2 (Optimal)

```text
10 → 30 → 20

Cost:
|10−30| = 20  
|30−20| = 10  

Total = 30  ✅ Minimum
```

---

## Recursive Logic

At each index:

Two choices:

```text
Jump 1 step → idx + 1  
Jump 2 steps → idx + 2
```

Take minimum:

```text
min(op1, op2)
```

Where:

```text
op1 = |h[i] − h[i+1]| + best(i+1)
op2 = |h[i] − h[i+2]| + best(i+2)
```

---

## Algorithm Steps

* Start
* Define recursive function `best(h, n, idx)`
* If `idx == n-1` → return 0
* Calculate cost of 1-step jump
* If at second last index → return op1
* Calculate cost of 2-step jump
* Return minimum of both
* Stop

---

## Python Code

```python
def best(h, n, idx):

    # Base case: reached last stone
    if idx == n - 1:
        return 0

    # Jump 1 step
    op1 = abs(h[idx] - h[idx + 1]) + best(h, n, idx + 1)

    # If second last index
    if idx == n - 2:
        return op1

    # Jump 2 steps
    op2 = abs(h[idx] - h[idx + 2]) + best(h, n, idx + 2)

    # Return minimum energy
    return min(op1, op2)


# Main execution
h = [10, 30, 40, 20]

print(best(h, len(h), 0))
```


# Python Problem: Keypad Combinations (Recursion)

## Problem Statement

Write a Python program to **print all possible letter combinations** for a given **digit string** based on a **mobile keypad** using **recursion**.

Each digit maps to a set of characters like a mobile keypad.

---

## Input

```text id="k6m6dp"
23
```

---

## Output

```text id="l4y8dq"
ad ae af bd be bf cd ce cf
```

*(Output may appear in same line separated by spaces)*

---

## Keypad Mapping

```text id="b3n9mu"
0 → ""
1 → ""
2 → "abc"
3 → "def"
4 → "ghi"
5 → "jkl"
6 → "mno"
7 → "pqr"
8 → "stuv"
9 → "wxyz"
```

---

## Explanation

For input:

```text id="f9b0xw"
"23"
```

Digit mappings:

```text id="x6d5s2"
2 → abc
3 → def
```

Possible combinations:

```text id="5b8kt9"
ad
ae
af
bd
be
bf
cd
ce
cf
```

Each character from first digit combines with all characters of second digit.

---

## Recursive Logic

At each step:

* Take first digit
* Find corresponding letters
* Recursively process remaining digits
* Append characters to result

---

## Algorithm Steps

* Start
* Define keypad mapping list
* Define recursive function `combination(dig, kp, res)`
* If digit string empty → print result
* Convert first digit to number
* Get corresponding characters
* Loop through characters
* Recursively call for remaining digits
* Stop

---

## Python Code

```python id="cobc9h"
def combination(dig, kp, res):

    # Base case
    if len(dig) == 0:
        print(res, end=" ")
        return

    # Convert first digit to number
    currNum = int(dig[0])

    # Get characters for digit
    currChoices = kp[currNum]

    # Try all possible characters
    for ch in currChoices:
        combination(dig[1:], kp, res + ch)


# Main execution
s = input()

keypad = [
    "", "", "abc", "def",
    "ghi", "jkl", "mno",
    "pqr", "stuv", "wxyz"
]

combination(s, keypad, "")
```
