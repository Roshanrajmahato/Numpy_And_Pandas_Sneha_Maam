# 3. First and Last Occurrence of an Element

### Given a sorted array and a target value, find the first and last index of the target using binary search techniques. If not found, return -1 -1.

![Alt text](<CodeMindMappImage\ChatGPT Image Mar 26, 2026, 04_28_09 PM.png>)

````python
def getFirst(v, target):
    n = len(v)
    lo = 0
    hi = n - 1
    ans = -1
    
    while lo <= hi:
        # mid = lo + (hi - lo) / 2
        mid = lo + (hi - lo) // 2
        
        if v[mid] == target:
            ans = mid
            hi = mid - 1  # Look on the left side for an earlier occurrence
        elif v[mid] < target:
            lo = mid + 1
        else:
            hi = mid - 1
            
    return ans

def getSecond(v, target):
    n = len(v)
    lo = 0
    hi = n - 1
    ans = -1
    
    while lo <= hi:
        mid = lo + (hi - lo) // 2
        
        if v[mid] == target:
            ans = mid
            lo = mid + 1  # Look on the right side for a later occurrence
        elif v[mid] < target:
            lo = mid + 1
        else:
            hi = mid - 1
            
    return ans


n = int(input("Enter n: "))

# vector<int> v(n)
v = []
print(f"Enter {n} sorted numbers:")
for _ in range(n):
    v.append(int(input()))
    
# cin >> target
target = int(input("Enter target: "))

first = getFirst(v, target)
second = getSecond(v, target)

# cout << "First : " << first << " Second : " << second;
print(f"First : {first} Second : {second}")
        
````

# Python Problem: Check Sphenic Number

## Problem Statement
Write a Python program to check whether a given number is a **Sphenic Number**.  
A number is called a **Sphenic number** if it is the product of **exactly three distinct prime numbers**.

**Input**
`n = 30`

**Output**
`Sphenic Number`

## Explanation
Prime factors of `30` → `2, 3, 5`  
Product → `2 × 3 × 5 = 30`  
Since the number has **exactly three distinct prime factors**, it is a **Sphenic Number**.

Example of Not Sphenic:  
`n = 60`  
Prime factors → `2, 2, 3, 5`  
It has more than three prime factors → **Not Sphenic**

````python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

n = int(input("Enter number: "))
primes = []

for i in range(2, n):
    if is_prime(i) and n % i == 0:
        primes.append(i)

if len(primes) == 3 and primes[0]*primes[1]*primes[2] == n:
    print("Sphenic Number")
else:
    print("Not Sphenic Number")
````

# 2️⃣ Check Whether List is Palindrome

From your notes:

* `left_idx = 0`
* `right_idx = len(l) - 1`
* Compare both sides

```python
# Check palindrome list

L = [10, 20, 30, 20, 10]

left_idx = 0
right_idx = len(L) - 1

while left_idx < right_idx:

    if L[left_idx] != L[right_idx]:
        print("List is not palindrome")
        break
    left_idx += 1
    right_idx -= 1

else:
    print("List is palindrome")
```
